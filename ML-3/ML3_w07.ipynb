```python
from sklearn.model_selection import learning_curve
from sklearn.linear_model import LinearRegression
import numpy as np
import matplotlib.pyplot as plt

# Пример данных
X = np.random.rand(100, 1)
y = 2 * X.ravel() + np.random.randn(100) * 0.1

# Вычисление кривых обучения
train_sizes, train_scores, val_scores = learning_curve(
    LinearRegression(), X, y, cv=5, scoring='neg_mean_squared_error',
    train_sizes=np.linspace(0.1, 1.0, 10)
)

# Средние значения и стандартное отклонение
train_scores_mean = -train_scores.mean(axis=1)  # Отрицательное, т.к. scoring='neg_...'
val_scores_mean = -val_scores.mean(axis=1)

# Построение графика
plt.plot(train_sizes, train_scores_mean, label='Train Error')
plt.plot(train_sizes, val_scores_mean, label='Validation Error')
plt.xlabel('Training Size')
plt.ylabel('MSE')
plt.legend()
plt.show()
```
