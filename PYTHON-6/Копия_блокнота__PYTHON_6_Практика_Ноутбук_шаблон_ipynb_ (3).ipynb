### Задание №1

#### Найти пересечение без дубликатов

**Компания:** «Авито»

**Уровень:** Лёгкий

___

Пользователь с клавиатуры (через функцию `input()`) вводит через запятую и пробел две последовательности чисел (некоторые идентификаторы товаров/услуг, просмотренных клиентом 'A' и клиентом 'B' на платформе «Авито»).

Необходимо конвертировать каждый ввод пользователя в отдельный список, состоящий из целых чисел, а также составить новый список, состоящий из элементов пересечения двух исходных. Дубликаты необходимо удалить, а результирующий список — вывести на экран.

**Важно!** Если пересечение пустое, то на экран должен выводиться пустой список.

Если ввод пользователя невозможно конвертировать в список из целых чисел, возникает ошибка. Необходимо отлавливать эту ошибку и выводить на экран фразу `Некорректный ввод`.

Для поиска пересечений в списках напишите функцию `pure_intersection()`, которая будет возвращать список с уникальными элементами пересечения.


**Пример №1:**
```
Ввод:
108, 138, 42, 52, 14
109, 13, 52, 32, 42, 14, 109

Вывод:
[42, 52, 14]

```

**Пример №2:**
```
Ввод:
108, 138, 42, 52, 14
189, 202, 249, 24, 15, 3

Вывод:
[]

```

**Пример №3:**
```
Ввод:
108, 138, 42, 52, четырнадцать
189, 202, 249, 24, 15, 3

Вывод:
Некорректный ввод

```

**Подсказка:**

Для определения пересечения преобразуйте списки во множества и воспользуйтесь методом `intersection()`.

**Критерии оценивания:**

- Задача решена корректно, выполнены все условия (**2 балла**):
    - каждый ввод пользователя конвертируется в список из идентификаторов;
    - функция для определения пересечения корректно работает для любого ввода пользователя, обозначенного в условиях задачи;
    - решение сопровождено комментариями.
- Задача решена корректно, однако не выполнено одно из условий (**1 балл**):
    - ввод пользователя не конвертируется в список;
    - функция для определения пересечения работает только для ограниченного множества входных данных в рамках условия задачи;
    - решение не сопровождено комментариями.
- Задача решена с серьёзными ошибками/не решена (**0 баллов**): представлен некорректный алгоритм определения пересечения элементов.

````python
input_string1 = input('Введите 1-ую последовательность идентификаторов: ')
input_string2 = input('Введите 2-ую последовательность идентификаторов: ')
# ваш код здесь

def pure_intersection(input_string1, input_string2):
    '''Функция возвращает найденное пересечение последовательностей, очищенных от дубликатов''' # Docstrings
    # каждый ввод пользователя конвертируется в список из идентификаторов
    result_list1 = list(map(lambda x: int(x), input_string1.split(', ')))
    result_list2 = list(map(lambda x: int(x), input_string2.split(', ')))
    # новый список, состоящий из элементов пересечения двух исходных списков, преобразованных в множества и очищенных от дубликатов
    result_list = list(set(result_list1).intersection(set(result_list2)))
    return result_list

# Вывод результатов
try: # обрабатываем исключения. Если ввод пользователя невозможно конвертировать в список из целых чисел - выводится 'Некорректный ввод'
    print(pure_intersection(input_string1, input_string2))
except:
    print('Некорректный ввод')
````

````python
input_string = input('Введите последовательность чисел: ')
# ваш код здесь

def find_min_max(input_string):
    '''Функция возвращает минимум и максимум последовательности из введенных пользователем действительных чисел, очищенной от ошибок при вводе''' # Docstrings
    input_list = input_string.replace(',', '.').split(' ') # заменяем в введенной строке ',' на '.' и разбиваем строку на подстроки
    output_list = [] # инициализируем результирующий список
    for index, elem in enumerate(input_list):
        try:
            if type(float(elem)) == float:
                output_list.append(float(elem)) # преобразуем строковый элемент списка в float и добавляем в новый список
            elif type(int(elem)) == int:
                output_list.append(int(elem))   # преобразуем строковый элемент списка в int и добавляем в новый список
        except:
            pass # пропускаем любой(!) элемент, который не можем преобразовать в int или float
    return min(output_list), max(output_list) # возвращаем минимальное и максимальное значение из результирующего списка

minimum, maximum = find_min_max(input_string)

# Вывод результатов
print(f'Minimum: {minimum}')
print(f'Maximum: {maximum}')
````

**Компания:** «Авито»

Медиана, как и среднее, является одной из важнейших характеристик для числовых данных в статистике.

**Медиана** — это число, которое находится в середине упорядоченной по возрастанию числовой последовательности.

Например, в последовательности чисел `1, 5, 2, 3, 6` медианой является число `3`, так как если упорядочить числа по возрастанию (`1, 2, 3, 5, 6`), центральным элементом в последовательности будет именно `3`.

**Примечание.** *Если в выборке чётное число элементов, то за медиану принимается среднее между центральными элементами (после упорядочивания).*

Например, в последовательности чисел `100, 5, 2, 4, 3, 6` медианой является число `4.5`, так как, если упорядочить числа по возрастанию (`2, 3, 4, 5, 6, 100`), то центральным элементом в последовательности будет среднее арифметическое между 4 и 5 ($\frac{4 + 5} {2} = 4.5$).

---

````python
input_string = input('Введите последовательность чисел: ')
# ваш код здесь

def find_median(input_string):
    '''Функция возвращает медиану из последовательности из введенных пользователем действительных чисел, очищенной от ошибок при вводе''' # Docstrings
    # в задании не сказано, что пользователь вводит только целые числа, поэтому предполагаем возможность обработки как int, так и float
    if not input_string: # если переданная в функцию последовательность окажется пустой (пользователь ничего не ввёл)
            print('Некорректный ввод')
            return
    input_list = input_string.split(', ') # разбиваем строку на подстроки и преобразуем в список
    output_list = [] # инициализируем результирующий список
    for index, elem in enumerate(input_list):
        try:
            if type(float(elem)) == float:
                output_list.append(float(elem)) # преобразуем строковый элемент списка в float и добавляем в новый список
            elif type(int(elem)) == int:
                output_list.append(int(elem)) # преобразуем строковый элемент списка в int и добавляем в новый список
        except:
            print('Некорректный ввод') # если встретился элемент, который не можем преобразовать в int или float
            return
    if not output_list: # если выходной список с числами - пустой
        print('Некорректный ввод')
        return
    output_list.sort() # сортируем выходной числовой список
    if len(output_list)%2 ==0:
        mediana = (output_list[len(output_list)//2-1]+output_list[len(output_list)//2])/2 # если в результирующем списке четное число значений
    else:
        mediana = output_list[len(output_list)//2] # если в результирующем списке нечетное число значений

    return mediana # возвращаем значение медианы из результирующего списка

mediana = find_median(input_string)

# Вывод результатов
print(f'Median: {mediana}' if mediana != None else '')
````

**Компания:** «СБЕР»

**Уровень:** Сложный

___

Напишите функцию `decompose_factorial()`, которая будет принимать на вход введённое пользователем число `n`, а затем вычислять факториал этого числа и выполнять разложение факториала на простые множители.

Напомним, что **факториал числа $n$ ($n!$)** — это произведение всех натуральных чисел от $1$ до $n$ включительно.

**Простое число** — число, которое делится (нацело) только на себя и на 1. Например, $13$ — простое число, так как оно делится только на себя (на $13$) и на $1$, а $14$ не является простым числом, так как оно делится на $7$ и на $2$.

Подробнее о простых числах можно почитать [здесь](https://dzen.ru/a/XqFMe_VWclT8NTss).

---
**Простые множители** — это простые числа (не считая 1), на которые число делится без остатка.

Например, для числа $10$ простыми множителями являются числа $5$ и $2$. Для числа $28$ множителями являются числа $7$ и $4$, но простыми множителями будут числа $7$ и $2$, так как $4$ — это $2 * 2$, а число $28$ представляется в виде $28 = 7 * 2^2$.

----

В результате своей работы функция должна возвращать строку вида (приведён пример для `n=12`):

`12! = 2^10 * 3^5 * 5^2 * 7 * 11`

Здесь число слева от ^ — множитель, число справа от ^ — его степень. При этом если множитель встречается один раз, знак степени не выводится. Вывод множителей должен производиться в порядке их возрастания.

````python
n = input('Введите число: ')
# ваш код здесь

def decompose_factorial(n):
    '''Функция принимает на вход число n, вычислять факториал этого числа и выполняет разложение факториала на простые множители''' # Docstrings

    def factorial(n):
        '''Функция вычислять факториал  от аргумента n ''' # Docstrings
        if n == 1 or n == 0:
            return 1
        return n * factorial(n-1)

    def factor(n):
        '''Функция реализует разложение числа n на множители''' # Docstrings
        Ans = []
        d = 2
        while d * d <= n:
            if n % d == 0:
                Ans.append(d)
                n //= d
            else:
                d += 1
        if n > 1:
            Ans.append(n)
        return Ans

    result_list = factor(factorial(n)) # список множителей после разложения факториала
    result_set = set(result_list) # находим множители без повторов - в виде множества
    result_dict =  dict(sorted(dict.fromkeys(result_set, 0).items() )) # формируем отсортированный словарь множителей, значения = степень данного множителя в разложении, заполняем значения нулями.

    # формируем словарь из множителей, составляющих разложенный факториал. Ключ - множитель, значение - его степень (количество)
    for key in result_dict:
        for multiplier in result_list:
            if multiplier == key:
                result_dict[key] +=1
    # возвращаем результат
    return result_dict

# добавлена проверка ввода пользователя, в задании не оговорена
try:
    result = decompose_factorial(int(n))

    # Вывод в заданном формате
    print(f'{n}! = ', end = '')
    for keys,values in result.items():
        if keys != max(result.keys()): # вывод последнего элемента не должен содержать завершающий знак умножения
            multy_sign = ' * '
        else:
            multy_sign = ''

        if values !=1:
            print(f'{keys}^{values}{multy_sign}', end = '')
        else:
            print(f'{keys}{multy_sign}', end = '')

    print() # Завершаем строку вывода
except (TypeError, ValueError):
    print('Некорректный ввод')
except RecursionError:
    print('Слишком большое число для рекурсивного расчете факториала')

````

## Задание №6 (бонусная, не оценивается)

#### Заменить дублирующиеся буквы

**Компания:** «СБЕР»

**Уровень:** Сложный

____

Пользователь с клавиатуры (через функцию `input()`) вводит строку, состоящую из символов латинского алфавита.

Необходимо последовательно заменять две одинаковые буквы во введённой строке на следующую по алфавиту букву (важно: заменяем по 2 одинаковых буквы за одну итерацию, не более). Операция замены должна повторяться до тех пор, пока в строке не останутся только уникальные символы.

**Несколько важных моментов:**

* буквы не обязаны быть смежными (смежные — стоящие рядом; в строке `'abc'` буквы a и b, а также b и c являются смежными, в то время как a и c — нет);
* zz заменяем на a;
* вводимая строка состоит из букв латинского алфавита в нижнем регистре.

Напишите функцию `replace_duplicates()`, которая принимает на вход произвольную строку `s`, значение которой вводится с клавиатуры. Функция должна последовательно производить описанную операцию и возвращать строку с оставшимися буквами в любом порядке.

**Пример:**
```
input:

step 1: zaba
step 2: zbb

output:
cz
```


**Подсказка:**

При решении задачи вам может понадобиться латинский алфавит. Вы можете получить его из встроенной в Python библиотеки `string`:

````python
import string
alphabet = string.ascii_lowercase
print(alphabet)

````

````python
# ваш код здесь
s = input('Введите строку: ')

def replace_duplicates(string):
    '''Функция последовательно заменяет две одинаковые буквы во введённой строке на следующую по алфавиту букву''' # Docstrings

    def next_symbol(substr): # вложенная функция подбора символа, заменяющего дубликаты
        next = alphabet.index(substr)+1
        if next >= len(alphabet):
            next = 0
        return alphabet[next]

    def replace_substring(string_list): # вложенная функция поиска 2 дублирующихся символов и их замены на 1 новый символ
        for i1, symbol1 in enumerate(string_list):
            for i2, symbol2 in enumerate(string_list[i1+1:]):
                if symbol1 == symbol2:
                    pos1 = i1
                    pos2 = i1+1+i2
                    string_list.pop(pos2)
                    string_list.pop(pos1)
                    string_list.append(next_symbol(symbol1))

                    return string_list
        return string_list

    i_step = 1 # номер шага последовательной операции замены
    string_list = list(string) # преобразуем строку в список

    while len(string_list) != len(set(string_list)): # цикл итераций по строке, условие выхода из цикла - в строке отсутствуют дубликаты
        string_list = replace_substring(string_list)
        print('step{}: {}'.format(i_step, ''.join(string_list))) # В образце решения есть вывод итогов каждой последовательной операции
        i_step +=1
    # возвращаем результат
    return ''.join(string_list)


#вывод
try: # добавлена проверка ввода пользователя, в задании не оговорена
    if len(set(list(s)).union(set(alphabet))) != len(set(alphabet)):
        raise ValueError("Несоответствующее значение")
    print()
    result = replace_duplicates(s)
    print()
    print('output:')
    print(result) # выводим строку с оставшимися буквами
except:
    print('Некорректный ввод')
````
