### Задача 3.1

1/1 point (graded)


Найдите значение формулы  $\neg(x) \vee(\neg(y) \vee z)$  при  $x=0$ ,  $y=1$  и  $z=0$ .

Будет ли значение этой формулы отличаться от значения  $(\neg(x) \vee \neg(y)) \vee z$  на том же наборе значений переменных?

### Решение: 

Выполним вычисления для каждой формулы по шагам.

**Дано:**
$x = 0$
$y = 1$
$z = 0$

**1. Вычислим значение первой формулы:**
$$\neg (x) \lor (\neg (y) \lor z)$$

*   Подставляем значения: $\neg (0) \lor (\neg (1) \lor 0)$
*   Вычисляем отрицания: $1 \lor (0 \lor 0)$
*   Вычисляем выражение в скобках: $0 \lor 0 = 0$
*   Итоговое действие: $1 \lor 0 = 1$

Значение первой формулы равно **1**.

**2. Вычислим значение второй формулы:**
$$(\neg (x) \lor \neg (y)) \lor z$$

*   Подставляем значения: $(\neg (0) \lor \neg (1)) \lor 0$
*   Вычисляем отрицания: $(1 \lor 0) \lor 0$
*   Вычисляем выражение в скобках: $1 \lor 0 = 1$
*   Итоговое действие: $1 \lor 0 = 1$

Значение второй формулы равно **1**.

**3. Ответ на вопрос:**
Вопрос звучит так: *«Будет ли значение этой формулы отличаться от значения... на том же наборе значений переменных?»*

Так как $1 = 1$, значения формул совпадают, то есть **не отличаются**.
(Это также следует из свойства ассоциативности дизъюнкции: порядок скобок в выражении $A \lor B \lor C$ не важен).

**Ответ:**
**Нет**

### Задача 3.2

1/1 point (graded)

Какие из следующих формул являются тавтологиями (то есть какие бы булевы значения не подставлять вместо переменных, значение всегда будет равно 1)?

**A** $x \to y$

**B** $x \to (y \land x)$

**C** $x \land (\neg x)$

**D** $x \to (\neg x \to y)$


### Решение:

Для того чтобы определить, какие формулы являются тавтологиями, нужно проверить, всегда ли они принимают значение **1** (Истина) при любых значениях переменных $x$ и $y$.

Рассмотрим каждый вариант:

1.  **A** $x \to y$
    *   Это обычная импликация. Она ложна (равна 0), если $x = 1$, а $y = 0$.
    *   Так как существует случай, когда формула равна 0, это **не тавтология**.

2.  **B** $x \to (y \land x)$
    *   Раскроем импликацию через формулу $\neg A \lor B$:
        $\neg x \lor (y \land x)$
    *   Применим закон дистрибутивности: $(\neg x \lor y) \land (\neg x \lor x)$.
    *   Так как $(\neg x \lor x)$ всегда равно 1, выражение упрощается до $(\neg x \lor y)$, что эквивалентно $x \to y$.
    *   Как мы выяснили в пункте А, это выражение может быть ложным (при $x=1, y=0$).
    *   Значит, это **не тавтология**.

3.  **C** $x \land (\neg x)$
    *   Это логическое противоречие. Переменная не может быть одновременно и истинной, и ложной.
    *   Если $x=1$, то $1 \land 0 = 0$.
    *   Если $x=0$, то $0 \land 1 = 0$.
    *   Формула всегда равна 0. Это **не тавтология**.

4.  **D** $x \to (\neg x \to y)$
    *   Раскроем внутреннюю импликацию: $(\neg x \to y)$ эквивалентно $(\neg(\neg x) \lor y)$, то есть $(x \lor y)$.
    *   Теперь подставим это в исходную формулу: $x \to (x \lor y)$.
    *   Раскроем внешнюю импликацию: $\neg x \lor (x \lor y)$.
    *   Сгруппируем слагаемые (дизъюнкция ассоциативна): $(\neg x \lor x) \lor y$.
    *   Выражение $(\neg x \lor x)$ всегда равно **1** (закон исключённого третьего).
    *   Выражение $1 \lor y$ всегда равно **1**.
    *   Следовательно, формула истина при любых значениях переменных. Это **тавтология**.

**Ответ:**
**D** $x \to (\neg x \to y)$

### Задача 4.2

1/1 point (graded)

Дана булева формула  $x_1 \rightarrow\left(x_2 \rightarrow\left(x_3 \rightarrow\left(\ldots \rightarrow\left(x_{10}\right) \right)\right)\right.$ , то есть 9 вложенных импликаций. Сколько имеется значений, на которых эта формула принимает значение 0?


### Решение: 

Проведем анализ структуры логической формулы.

**Дано:**
Формула с 9 вложенными импликациями:
$$ x_1 \to (x_2 \to (x_3 \to (\dots \to (x_{10})\dots))) $$

**Вопрос:**
На скольких наборах значений переменных эта формула принимает значение **0** (Ложь)?

**Решение:**

1.  **Свойство импликации:**
    Операция импликации $A \to B$ равна **0** (Лжи) тогда и только тогда, когда:
    *   $A = 1$ (Истина)
    *   $B = 0$ (Ложь)
    Во всех остальных случаях импликация равна 1.

2.  **Применим это к нашей формуле:**
    Обозначим $F = x_1 \to (x_2 \to (\dots))$.
    Чтобы $F = 0$, необходимо:
    *   Левая часть: **$x_1 = 1$**
    *   Правая часть: $(x_2 \to (\dots)) = 0$

3.  **Рекурсивный спуск:**
    Теперь рассматриваем правую часть. Чтобы выражение $(x_2 \to (x_3 \to \dots))$ было равно 0, снова применяем правило импликации:
    *   **$x_2 = 1$**
    *   Остаток $(x_3 \to \dots) = 0$

    Повторяя это рассуждение шаг за шагом, мы получаем однозначные требования для всех переменных по цепочке:
    *   $x_3 = 1$
    *   ...
    *   $x_8 = 1$
    *   Остаток: $(x_9 \to x_{10}) = 0$

4.  **Последний шаг:**
    Чтобы последняя импликация $x_9 \to x_{10}$ была равна 0, должно выполняться:
    *   **$x_9 = 1$**
    *   **$x_{10} = 0$**

5.  **Итог:**
    Формула принимает значение 0 только в **одном единственном случае**, когда переменные имеют следующие значения:
    $$ x_1=1, x_2=1, x_3=1, \dots, x_9=1, x_{10}=0 $$

Никакой другой комбинации не существует. Если хотя бы одна переменная от $x_1$ до $x_9$ будет равна 0, соответствующая импликация станет равна 1, и по цепочке "наверх" всё выражение станет истинным. Если $x_{10}$ будет равно 1, то последняя скобка станет 1, и всё выражение тоже станет 1.

**Ответ:** 1

### Задача 4.5

1/1 point (graded)

На первом году существования компании в ней был набран в штат один человек. Каждый следующий год набирается в два раза больше сотрудников, чем было набрано годом ранее.

Сколько сотрудников было набрано за всё время существование компании за 20 лет?

**Ответ:** 1048575

Для решения задачи нужно подсчитать сумму геометрической прогрессии.

**1. Анализ условия:**
*   **1-й год:** нанят $1$ сотрудник.
*   **2-й год:** нанято в 2 раза больше, то есть $1 \times 2 = 2$ сотрудника.
*   **3-й год:** нанято $2 \times 2 = 4$ сотрудника.
*   **4-й год:** нанято $4 \times 2 = 8$ сотрудников.

Мы видим последовательность чисел: $1, 2, 4, 8, \dots, 2^{n-1}$.
Это геометрическая прогрессия, где:
*   Первый член ($b_1$) = $1$
*   Знаменатель прогрессии ($q$) = $2$
*   Количество лет ($n$) = $20$

**2. Формула суммы:**
Нам нужно найти общее количество сотрудников за 20 лет, то есть сумму первых 20 членов этой прогрессии ($S_{20}$).
Формула суммы геометрической прогрессии:
$$ S_n = \frac{b_1 \cdot (q^n - 1)}{q - 1} $$

Подставим наши значения:
$$ S_{20} = \frac{1 \cdot (2^{20} - 1)}{2 - 1} = 2^{20} - 1 $$

**3. Вычисление:**
Нужно вычислить $2^{20} - 1$.
Известно, что $2^{10} = 1024$.
Тогда $2^{20} = 2^{10} \times 2^{10} = 1024 \times 1024$.

Произведем умножение:
$1024 \times 1024 = 1\ 048\ 576$

Вычтем единицу:
$1\ 048\ 576 - 1 = 1\ 048\ 575$


Мы использовали формулу для частного случая геометрической прогрессии ($b_1 = 1, q = 2$). Докажем методом математической индукции утверждение:

$$ \sum_{i=0}^{n-1} 2^i = 1 + 2 + 4 + \dots + 2^{n-1} = 2^n - 1 $$


### Доказательство методом математической индукции

**1. База индукции (Base case)**
Проверим истинность утверждения для $n = 1$ (первый год работы компании).
*   **Левая часть (сумма):** В первый год нанят $2^{1-1} = 2^0 = 1$ сотрудник.
*   **Правая часть (формула):** $2^1 - 1 = 2 - 1 = 1$.
*   $1 = 1$.
*   **Вывод:** Для $n=1$ утверждение верно.

**2. Индуктивное предположение (Inductive hypothesis)**
Предположим, что формула верна для некоторого числа лет $k$. То есть мы считаем истинным равенство:
$$ 1 + 2 + \dots + 2^{k-1} = 2^k - 1 $$

**3. Индуктивный переход (Inductive step)**
Нужно доказать, что формула будет верна и для следующего года $n = k + 1$.
Нам нужно найти сумму для $k+1$ лет:
$$ S_{k+1} = \underbrace{(1 + 2 + \dots + 2^{k-1})}_{\text{Сумма за } k \text{ лет}} + \underbrace{2^{(k+1)-1}}_{\text{Найм в } (k+1)\text{-й год}} $$

*   Вместо суммы за $k$ лет подставим наше **индуктивное предположение** ($2^k - 1$).
*   Число сотрудников, нанятых в $(k+1)$-й год, равно $2^k$.

Получаем:
$$ S_{k+1} = (2^k - 1) + 2^k $$

Сгруппируем слагаемые:
$$ S_{k+1} = 2^k + 2^k - 1 $$
$$ S_{k+1} = 2 \cdot 2^k - 1 $$
$$ S_{k+1} = 2^{k+1} - 1 $$

**Вывод:**
Мы получили, что формула для $k+1$ лет ($2^{k+1} - 1$) выводится из формулы для $k$ лет.
Согласно принципу математической индукции, утверждение верно для любого натурального числа $n$.

### Задача 4.6

1/1 point (graded)

Докажите по индукции, что сумма квадратов первых  $n$  натуральных чисел равна  $\frac{n(n+1)(2 n+1)}{6}$ . С помощью этой формулы найдите сумму первых 100 квадратов натуральных чисел.

**Ответ:** 338350

### Решение: 

Решение задачи выполним два этапа: сначала докажем формулу методом математической индукции, а затем вычислим требуемую сумму.

#### Часть 1. Доказательство индукцией

**Утверждение:** $1^2 + 2^2 + \dots + n^2 = \frac{n(n+1)(2n+1)}{6}$ для любого натурального $n$.

**1. База индукции ($n = 1$):**
*   Левая часть (сумма первого квадрата): $1^2 = 1$.
*   Правая часть (формула): $\frac{1(1+1)(2\cdot1+1)}{6} = \frac{1 \cdot 2 \cdot 3}{6} = \frac{6}{6} = 1$.
*   $1 = 1$. База верна.

**2. Индуктивное предположение:**
Предположим, что формула верна для некоторого $k$:
$$1^2 + \dots + k^2 = \frac{k(k+1)(2k+1)}{6}$$

**3. Индуктивный переход ($n = k+1$):**
Нужно доказать, что формула верна для $k+1$.
Прибавим к обеим частям равенства из предположения следующее слагаемое $(k+1)^2$:
$$1^2 + \dots + k^2 + (k+1)^2 = \frac{k(k+1)(2k+1)}{6} + (k+1)^2$$

Преобразуем правую часть, чтобы она приняла вид формулы для $k+1$:
1.  Вынесем общий множитель $(k+1)$ за скобки:
    $$ (k+1) \left( \frac{k(2k+1)}{6} + (k+1) \right) $$
2.  Приведем выражение в скобках к общему знаменателю:
    $$ (k+1) \left( \frac{2k^2 + k + 6(k+1)}{6} \right) = (k+1) \left( \frac{2k^2 + k + 6k + 6}{6} \right) $$
3.  Упростим числитель:
    $$ (k+1) \frac{2k^2 + 7k + 6}{6} $$
4.  Разложим квадратный трехчлен $2k^2 + 7k + 6$ на множители.
    Дискриминант $D = 49 - 4\cdot2\cdot6 = 1$. Корни: $k_1 = -2, k_2 = -1.5$.
    Значит, $2k^2 + 7k + 6 = 2(k+2)(k+1.5) = (k+2)(2k+3)$.
5.  Подставим обратно:
    $$ \frac{(k+1)(k+2)(2k+3)}{6} $$

Заметим, что полученное выражение — это в точности исходная формула, где вместо $n$ подставлено $k+1$:
$$ \frac{(k+1)((k+1)+1)(2(k+1)+1)}{6} $$
Доказательство завершено.


#### Часть 2. Вычисление суммы

Нам нужно найти сумму первых 100 квадратов ($n = 100$).
Подставим $n=100$ в доказанную формулу:

$$ S_{100} = \frac{100(100+1)(2 \cdot 100 + 1)}{6} $$

1.  Вычислим множители:
    $100 \cdot 101 \cdot 201$
2.  Подставим в дробь:
    $$ \frac{100 \cdot 101 \cdot 201}{6} $$
3.  Сократим дробь (201 делится на 3, 100 делится на 2):
    $\frac{100}{2} \cdot 101 \cdot \frac{201}{3} = 50 \cdot 101 \cdot 67$
4.  Произведем умножение:
    *   $50 \cdot 101 = 5050$
    *   $5050 \cdot 67 = 338350$

**Ответ:** 338350
