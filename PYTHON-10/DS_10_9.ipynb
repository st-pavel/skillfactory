Задание 9.12
1 point possible (graded)
Насколько медианный балл по письму у студентов в расовой группе А отличается от среднего балла по письму у студентов в расовой группе C?
Округлите ответ до целого и запишите модуль этого числа.

```python


student_race_a_writing_median = student_data[student_data['race/ethnicity'] == 'group A']['writing score'].median()
student_race_c_writing_mean = student_data[student_data['race/ethnicity'] == 'group C']['writing score'].mean()
display(student_race_a_writing_median)
display(student_race_c_writing_mean)
delta_score = abs(round(student_race_a_writing_median - student_race_c_writing_mean,0))
display(delta_score)
```
