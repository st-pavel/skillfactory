```python
# –∏–º–ø–æ—Ä—Ç–∏—Ä—É–µ–º –Ω—É–∂–Ω—ã–µ –±–∏–±–ª–∏–æ—Ç–µ–∫–∏
from itertools import permutations
import math
from collections import Counter  # –î–ª—è –ø–æ–¥—Å—á—ë—Ç–∞ –ø–æ–≤—Ç–æ—Ä—è—é—â–∏—Ö—Å—è –±—É–∫–≤

word = "–ò–ù–¢–ï–õ–õ–ï–ö–¢"
n = len(word)

# –ü–æ—Å—á–∏—Ç–∞–µ–º, —Å–∫–æ–ª—å–∫–æ —Ä–∞–∑ –ø–æ–≤—Ç–æ—Ä—è–µ—Ç—Å—è –∫–∞–∂–¥–∞—è –±—É–∫–≤–∞
letter_counts = Counter(word)

# –í—ã—á–∏—Å–ª–∏–º –∑–Ω–∞–º–µ–Ω–∞—Ç–µ–ª—å ‚Äî –ø—Ä–æ–∏–∑–≤–µ–¥–µ–Ω–∏–µ —Ñ–∞–∫—Ç–æ—Ä–∏–∞–ª–æ–≤ –∫–æ–ª–∏—á–µ—Å—Ç–≤ –ø–æ–≤—Ç–æ—Ä–µ–Ω–∏–π
denominator = 1
for count in letter_counts.values():
    denominator *= math.factorial(count)

# –û–±—â–µ–µ –∫–æ–ª–∏—á–µ—Å—Ç–≤–æ —É–Ω–∏–∫–∞–ª—å–Ω—ã—Ö –ø–µ—Ä–µ—Å—Ç–∞–Ω–æ–≤–æ–∫
total_permutations = math.factorial(n) // denominator

print(f"–ö–æ–ª–∏—á–µ—Å—Ç–≤–æ —É–Ω–∏–∫–∞–ª—å–Ω—ã—Ö –ø–µ—Ä–µ—Å—Ç–∞–Ω–æ–≤–æ–∫: {total_permutations}")
```

```python
# –∏–º–ø–æ—Ä—Ç–∏—Ä—É–µ–º –Ω—É–∂–Ω—ã–µ –±–∏–±–ª–∏–æ—Ç–µ–∫–∏
from math import comb

# –∫–æ–ª–∏—á–µ—Å—Ç–≤–æ —Ä–∞–∑–ª–∏—á–Ω—ã—Ö —Ç–∏–ø–æ–≤ –Ω–∞–ø–∏—Ç–∫–æ–≤ (n): –∫–æ—Ñ–µ, —á–∞–π, —Å–æ–∫, –≤–æ–¥–∞
n = 4 
# –∫–æ–ª–∏—á–µ—Å—Ç–≤–æ –Ω–∞–ø–∏—Ç–∫–æ–≤, –∫–æ—Ç–æ—Ä–æ–µ –Ω—É–∂–Ω–æ –∑–∞–∫–∞–∑–∞—Ç—å (k)
k = 7

# –∫–æ–ª–∏—á–µ—Å—Ç–≤–æ —É–Ω–∏–∫–∞–ª—å–Ω—ã—Ö —Å–ø–æ—Å–æ–±–æ–≤, –∫–æ—Ç–æ—Ä—ã–º–∏ –º—ã –º–æ–∂–µ–º –∑–∞–∫–∞–∑–∞—Ç—å 7 –Ω–∞–ø–∏—Ç–∫–æ–≤, –µ—Å–ª–∏ –º—ã –º–æ–∂–µ–º –≤—ã–±–∏—Ä–∞—Ç—å —Ä–∞–∑–Ω—ã–µ —Ç–∏–ø—ã, –º–æ–∂–Ω–æ —Ä–∞—Å—Å—á–∏—Ç–∞—Ç—å –∏—Å–ø–æ–ª—å–∑—É—è —Ñ–æ—Ä–º—É–ª—É C(n+k-1, k)
combinations_with_repetitions = comb(n + k - 1, k)
combinations_with_repetitions
```

```python
# –ü—Ä–æ–≥—Ä–∞–º–º–∞ –¥–ª—è –≤—ã–≤–æ–¥–∞ —Ç—Ä–µ—É–≥–æ–ª—å–Ω–∏–∫–∞ –ü–∞—Å–∫–∞–ª—è
from math import comb

n = int(input())  # –í–≤–æ–¥–∏–º –∫–æ–ª–∏—á–µ—Å—Ç–≤–æ —Å—Ç—Ä–æ–∫

# –í—ã–≤–æ–¥–∏–º –ø–µ—Ä–≤—ã–µ n —Å—Ç—Ä–æ–∫ —Ç—Ä–µ—É–≥–æ–ª—å–Ω–∏–∫–∞ –ü–∞—Å–∫–∞–ª—è
for i in range(n):
    # –§–æ—Ä–º–∏—Ä—É–µ–º i-—é —Å—Ç—Ä–æ–∫—É —Ç—Ä–µ—É–≥–æ–ª—å–Ω–∏–∫–∞ –ü–∞—Å–∫–∞–ª—è
    row = []
    for j in range(i + 1):
        # –ö–∞–∂–¥—ã–π —ç–ª–µ–º–µ–Ω—Ç - —ç—Ç–æ –±–∏–Ω–æ–º–∏–∞–ª—å–Ω—ã–π –∫–æ—ç—Ñ—Ñ–∏—Ü–∏–µ–Ω—Ç C(i, j)
        row.append(str(comb(i, j)))
    # –í—ã–≤–æ–¥–∏–º —Å—Ç—Ä–æ–∫—É –±–µ–∑ –ø—Ä–æ–±–µ–ª–æ–≤ –≤ –Ω–∞—á–∞–ª–µ, —ç–ª–µ–º–µ–Ω—Ç—ã —Ä–∞–∑–¥–µ–ª–µ–Ω—ã –ø—Ä–æ–±–µ–ª–∞–º–∏
    print(' '.join(row))
```

```python
n = 15

# –ü–µ—Ä–≤–∞—è —Å—Ç—Ä–æ–∫–∞ –≤—Å–µ–≥–¥–∞ [1]
row = [1]

for _ in range(n):
    # –í—ã–≤–æ–¥–∏–º —Ç–µ–∫—É—â—É—é —Å—Ç—Ä–æ–∫—É, —Ä–∞–∑–¥–µ–ª—è—è —ç–ª–µ–º–µ–Ω—Ç—ã –ø—Ä–æ–±–µ–ª–æ–º
    # –ó–≤–µ–∑–¥–æ—á–∫–∞ (*) —Ä–∞—Å–ø–∞–∫–æ–≤—ã–≤–∞–µ—Ç —Å–ø–∏—Å–æ–∫, —É–±–∏—Ä–∞—è –∫–≤–∞–¥—Ä–∞—Ç–Ω—ã–µ —Å–∫–æ–±–∫–∏ –∏ –∑–∞–ø—è—Ç—ã–µ
    print(*row)
    
    # –§–æ—Ä–º–∏—Ä—É–µ–º —Å–ª–µ–¥—É—é—â—É—é —Å—Ç—Ä–æ–∫—É
    # –õ–æ–≥–∏–∫–∞: [0] + row –∏ row + [0] —Å–¥–≤–∏–≥–∞—é—Ç —Å–ø–∏—Å–æ–∫ –æ—Ç–Ω–æ—Å–∏—Ç–µ–ª—å–Ω–æ –¥—Ä—É–≥ –¥—Ä—É–≥–∞
    # zip —Å–∫–ª–∞–¥—ã–≤–∞–µ—Ç —ç–ª–µ–º–µ–Ω—Ç—ã –ø–æ–ø–∞—Ä–Ω–æ: (0+1), (1+2), (2+1), (1+0) –∏ —Ç.–¥.
    row = [x + y for x, y in zip([0] + row, row + [0])]
```

```python
levels = int(input()) 

last_state = [1]
triangle = []

for count in range(levels):
    row = [1]
    for i in range(count):
        row.append(sum(last_state[i: i + 2]))
    last_state = row
    triangle.append(row)

for row in triangle:
    print(*row)
```
