# PROJECT-2. Анализ вакансий из HeadHunter

## 8. Подготовка к аттестации по SQL

✍ В следующем модуле вам предстоит пройти аттестацию по разделу SQL. Мы включили в тест теоретические вопросы и задачи на знание и умение применять SQL, которые вы можете встретить при устройстве на работу.

Все задания будут проверяться системой автоматически, так что вы сразу сможете узнать, верно ли вы ответили на вопрос или решили задачу.

Всего в аттестации будет 16 заданий: 11 теоретических вопросов и 5 заданий на написание запросов, для решениях которых вам предстоит поработать с небольшой базой данных и написать несколько запросов к ней. Данные для аттестационных заданий находятся в схеме sqlprotest из базы данных Skillfactory, которая доступна в Metabase.

Это немного, но достаточно для того, чтобы на собеседовании судить об уровне владения SQL.

На каждое задание даётся одна попытка. Обязательно проверяйте задания на написание кода в Metabase — это позволит вам избежать синтаксических ошибок. Также проверяйте результат на соответствие заданию.

Перед выполнением заданий на написание запросов рекомендуем внимательно ознакомиться с описанием и содержимым таблиц.

Обращайте внимание на подсказки к заданиям — это может помочь вам найти правильное решение.

⚡ ОЧЕНЬ ВАЖНО! На прохождение аттестации вам даётся 2 часа с момента, как вы приступили к заданиям. По истечении указанного времени отправить ответы уже не получится.

Повторите темы, в которых, как вы считаете, у вас есть пробелы, и приступайте к аттестации.

Уверены, вам будет несложно. Желаем удачи!

Внимание!

Минимальный порог прохождения аттестации — 50 %.

Если вы набрали от 50 до 75 %, вы показали хороший уровень владения темой.
Если ваш результат превышает 75 %, вы отлично усвоили все материалы темы.
ЧТО ДЕЛАТЬ, ЕСЛИ РЕЗУЛЬТАТ МЕНЬШЕ 50 %?
В таком случае аттестация будет считаться непройденной. Однако не отчаивайтесь — у вас есть ещё 2 попытки, чтобы исправить результат.

Чтобы воспользоваться следующей попыткой, напишите запрос в Support — ваша предыдущая попытка прохождения аттестации будет сброшена.

Студентам, которые не смогли преодолеть порог в 50 % за 3 попытки, будет предложена консультация с тьютором для обсуждения дальнейшей работы над курсом.
