```python
columns_to_change = ['cylinders', 'title_status', 'transmission', 'drive', 'size']
 
for column in columns_to_change:
 print('Число уникальных значений призкака {}: '.format(column), data[column].nunique())
```

```python
from sklearn.preprocessing import OneHotEncoder
 
one_hot_encoder = OneHotEncoder()
 
# 'учим' и сразу применяем преобразование к выборке, результат переводим в массив
data_onehot = one_hot_encoder.fit_transform(data[columns_to_change]).toarray() 
# запишем полученные названия новых колонок в отдельную переменную
column_names = one_hot_encoder.get_feature_names_out(columns_to_change)
print(column_names)
```

```python
# Объединяем датафреймы по горизонтали (добавляем столбцы)
data_new = pd.concat([data, data_onehot], axis=1)

print("\nПервые несколько строк:")
print(data_new.head())
```

```python
# Удаляем только столбцы, точно совпадающие с именами из списка
data_new = data_new.drop(columns=columns_to_change, errors='ignore')

# Параметр errors='ignore' позволяет пропускать столбцы, которых нет в датафрейме
# без возникновения ошибки

# Проверяем результат
print("Удаленные столбцы:", columns_to_change)
print("\nРазмерность после удаления:", data_new.shape)
print("\nОставшиеся столбцы:")
print(data_new.columns.tolist())

```
