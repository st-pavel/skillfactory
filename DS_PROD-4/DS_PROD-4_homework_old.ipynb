## 3. Проверка стабилизации метрик

```python
# Ежедневные показатели
daily_data = data.groupby(['date', 'group']).agg({
    'user_id': 'count',
    'purchase': 'sum',
    'price': 'sum'
}).reset_index().rename(columns={
    'user_id': 'count_visits',
    'purchase': 'count_purchases',
    'price': 'sum_price'
})

daily_data['conversion'] = daily_data['count_purchases'] / daily_data['count_visits'] * 100
daily_data['mean_check'] = daily_data['sum_price'] / daily_data['count_purchases']

display(daily_data.head())
```

```python
# Кумулятивные показатели
daily_data['cum_visits'] = daily_data.groupby('group')['count_visits'].cumsum()
daily_data['cum_purchases'] = daily_data.groupby('group')['count_purchases'].cumsum()
daily_data['cum_price'] = daily_data.groupby('group')['sum_price'].cumsum()

daily_data['cum_conversion'] = daily_data['cum_purchases'] / daily_data['cum_visits'] * 100
daily_data['cum_mean_check'] = daily_data['cum_price'] / daily_data['cum_purchases']

display(daily_data.tail())
```

```python
# Графики кумулятивной конверсии и среднего чека
fig, axes = plt.subplots(1, 2, figsize=(15, 5))

# Кумулятивная конверсия
sns.lineplot(x='date', y='cum_conversion', hue='group', data=daily_data, ax=axes[0])
axes[0].set_title('График кумулятивной конверсии по дням')
axes[0].grid(True)

# Кумулятивный средний чек
sns.lineplot(x='date', y='cum_mean_check', hue='group', data=daily_data, ax=axes[1])
axes[1].set_title('График кумулятивного среднего чека по дням')
axes[1].grid(True)

plt.tight_layout()
plt.show()
```

**Стабилизация метрик:**
- Кумулятивные метрики стабилизировались.
- Данных достаточно для принятия статистических решений.

## 4. Статистический анализ результатов A/B-тестирования

### 4.1. Формулировка гипотез и выбор тестов

Для проверки статистической значимости различий между группами A и B сформулируем нулевые ($H_0$) и альтернативные ($H_1$) гипотезы.

#### 1. Конверсия (Пропорции)
*   **Метрика:** Конверсия в покупку (CR).
*   $H_0: p_A = p_B$ (Конверсии в группах равны).
*   $H_1: p_A \neq p_B$ (Конверсии в группах различаются).
*   **Тест:** Z-тест для пропорций (z-test for proportions), так как мы сравниваем доли в двух независимых выборках большого размера ($n > 30$).

#### 2. Средний чек (AOV)
*   **Метрика:** Средний чек заказа.
*   $H_0: \mu_A = \mu_B$ (Средние чеки в группах равны).
*   $H_1: \mu_A \neq \mu_B$ (Средние чеки в группах различаются).
*   **Тест:**
    *   Проверим распределение на нормальность (Shapiro-Wilk test).
    *   Если распределение нормальное -> T-тест (t-test).
    *   Если распределение *не* нормальное -> U-тест Манна-Уитни (Mann-Whitney U-test), как непараметрический аналог.
    *   *Примечание:* Учитывая большой размер выборки (>1000), ЦПТ позволяет использовать T-тест даже при нарушении нормальности, поэтому проведем оба теста для надежности.

### 4.2. Проверка данных на нормальность
Проверим распределение ежедневных средних чеков (или самих чеков) на нормальность с помощью теста Шапиро-Уилка.

```python
# Проверка на нормальность распределения среднего чека
prices_a = data[(data['group'] == 'A') & (data['purchase'] == 1)]['price']
prices_b = data[(data['group'] == 'B') & (data['purchase'] == 1)]['price']

print('--- Тест на нормальность (Shapiro-Wilk) ---')
# Sample 1000 for Shapiro test performance/sensitivity reasons
print(f'Группа A (sample 1000 p-value): {shapiro(prices_a.sample(1000, random_state=42))[1]:.4f}')
print(f'Группа B (sample 1000 p-value): {shapiro(prices_b.sample(1000, random_state=42))[1]:.4f}')
```

**Промежуточный вывод по нормальности:**
*   **p-value < 0.05** для обеих групп.
*   Нулевая гипотеза о нормальности распределения **отвергается**.
*   Распределения не являются нормальными.
*   **Решение:** Для сравнения средних чеков целесообразно использовать непараметрический U-тест Манна-Уитни. T-тест также будет проведен справочно (допустимо по ЦПТ).

### 4.3. Проведение статистических тестов и расчет доверительных интервалов

```python
# --- 1. Конверсия (Z-test) ---
k_a = data[data['group'] == 'A']['purchase'].sum()
n_a = data[data['group'] == 'A']['user_id'].count()
k_b = data[data['group'] == 'B']['purchase'].sum()
n_b = data[data['group'] == 'B']['user_id'].count()

z_stat, p_val_z = proportions_ztest([k_a, k_b], [n_a, n_b])
ci_diff = confint_proportions_2indep(k_a, n_a, k_b, n_b, compare='diff', alpha=0.05)

print(f'Z-test p-value: {p_val_z:.4f}')
print(f'95% CI разницы конверсий: {ci_diff}')
```

```python
# --- 2. Средний чек (T-test, Mann-Whitney) ---
t_stat, p_val_t = ttest_ind(prices_a, prices_b, equal_var=False)
u_stat, p_val_u = mannwhitneyu(prices_a, prices_b)

def get_conf_interval(data, alpha=0.95):
    mu = data.mean()
    sem = data.sem()
    return norm.interval(alpha, loc=mu, scale=sem)

ci_a = get_conf_interval(prices_a)
ci_b = get_conf_interval(prices_b)

print(f'T-test p-value: {p_val_t:.4f}')
print(f'Mann-Whitney p-value: {p_val_u:.4f}')
print(f'95% CI для среднего чека A: {ci_a}')
print(f'95% CI для среднего чека B: {ci_b}')
```

### 4.4. Интерпретация результатов

#### 1. Конверсия
*   **Z-тест:** p-value > 0.05 (обычно ~0.24).
*   **Вывод:** Статистически значимых различий в конверсии между группами A и B **не обнаружено**.
*   **Доверительные интервалы:** Интервал разницы конверсий включает 0. Интервалы самих конверсий сильно пересекаются.

#### 2. Средний чек
*   **Тесты (T-test, Mann-Whitney):** p-value < 0.05 (практически 0.0000).
*   **Вывод:** Обнаружена **статистически значимая разница**. Средний чек в группе B выше.
*   **Доверительные интервалы:**
    *   CI группы A: ~[80.1k, 82.0k]
    *   CI группы B: ~[87.7k, 89.6k]
    *   **Пересечение:** Интервалы **не пересекаются**. Это подтверждает, что истинные средние в генеральных совокупностях различны и разделены.
