### Задание 3.4
1/1 point (graded)
Среди людей, заходящих в магазин, 70 % ничего не покупают. Остальные с равной вероятностью покупают одну-три бутылки воды. Найдите вероятность купить две бутылки воды, если известно, что количество купленных бутылок чётное.

Хорошо, давайте решим эту задачу по шагам.

**1. Определим основные события и их вероятности:**

*   Пусть событие `N` - человек ничего не покупает (0 бутылок).
    *   P(N) = 0.70
*   Пусть событие `B` - человек что-то покупает (1, 2 или 3 бутылки).
    *   P(B) = 1 - P(N) = 1 - 0.70 = 0.30
*   Пусть событие `B1` - человек покупает 1 бутылку.
*   Пусть событие `B2` - человек покупает 2 бутылки.
*   Пусть событие `B3` - человек покупает 3 бутылки.

По условию, если человек что-то покупает, то он покупает 1, 2 или 3 бутылки с равной вероятностью. Это означает, что вероятность покупки 1, 2 или 3 бутылок *условна* тем, что человек вообще что-то покупает (событие B).

*   P(B1 | B) = 1/3
*   P(B2 | B) = 1/3
*   P(B3 | B) = 1/3

Теперь найдем безусловные вероятности покупки 1, 2 или 3 бутылок:
*   P(B1) = P(B1 | B) * P(B) = (1/3) * 0.30 = 0.10
*   P(B2) = P(B2 | B) * P(B) = (1/3) * 0.30 = 0.10
*   P(B3) = P(B3 | B) * P(B) = (1/3) * 0.30 = 0.10

Проверим: P(N) + P(B1) + P(B2) + P(B3) = 0.70 + 0.10 + 0.10 + 0.10 = 1.00. Все верно.

**2. Определим события для условной вероятности:**

*   Пусть событие `A` - куплено ровно две бутылки воды (это то же самое, что событие B2).
    *   P(A) = P(B2) = 0.10
*   Пусть событие `E` - куплено чётное количество бутылок воды.
    *   Чётное количество - это 0 бутылок или 2 бутылки.
    *   P(E) = P(N) + P(B2) = 0.70 + 0.10 = 0.80

**3. Найдем искомую условную вероятность:**

Нам нужно найти вероятность купить две бутылки воды (событие A) при условии, что количество купленных бутылок чётное (событие E). То есть, ищем P(A | E).

Используем формулу условной вероятности:
P(A | E) = P(A ∩ E) / P(E)

*   `A ∩ E` - это событие, когда куплено ровно две бутылки (A) И куплено чётное количество бутылок (E). Если куплено ровно две бутылки, то автоматически куплено и чётное количество. Следовательно, событие `A ∩ E` совпадает с событием `A`.
*   P(A ∩ E) = P(A) = 0.10

Теперь подставляем значения в формулу:
P(A | E) = P(A) / P(E) = 0.10 / 0.80

**4. Вычисляем результат:**

P(A | E) = 0.10 / 0.80 = 10 / 80 = 1 / 8 = 0.125

**Ответ:** Вероятность купить две бутылки воды, если известно, что количество купленных бутылок чётное, равна 1/8 или 0.125.

*   Используем правило умножения вероятностей, которое вытекает из формулы условной вероятности P(P | F) = P(F ∩ P) / P(F).

**3. Найдите вероятность того, что случайный первокурсник SkillFactory получил повышение на работе в течение года.**

*   Нам нужно найти общую вероятность получения повышения P(P).

    P(Ora | Fail) = [ P(Fail | Ora) * P(Ora) ] / P(Fail)
*   Сначала найдем необходимые компоненты:
    *   P(Ora) = 0.35 (дано).
    *   P(Fail | Ora): Вероятность не сдать экзамен, если посещал курсы ораторского мастерства. Это дополнение к вероятности сдать:
        P(Fail | Ora) = 1 - P(Pass | Ora) = 1 - 0.75 = 0.25
    *   P(Fail): Общая вероятность не сдать экзамен. Это дополнение к общей вероятности сдать экзамен (P(Pass)), которую мы нашли в пункте 1:
        P(Fail) = 1 - P(Pass) = 1 - 0.6925 = 0.3075
*   Теперь подставляем значения в формулу Байеса:
    P(Ora | Fail) = (0.25 * 0.35) / 0.3075
    P(Ora | Fail) = 0.0875 / 0.3075
    P(Ora | Fail) ≈ 0.2845528...
*   Округляем до двух знаков после точки-разделителя: **0.28**

**Ответ на 2:** 0.28
