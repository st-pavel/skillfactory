### Задача 13

1 point possible (graded)

Сколько всего неизоморфных подграфов, порождённых четырьмя вершинами, у приведённого ниже графа?





### Решение

Исходя из приведенного рисунка:
- Вершина 1 соединена с 2, 3, 4, 5, 6, 8
- Вершина 2 соединена с 1, 2, 3, 5, 6, 7, 8
- Вершина 3 соединена с 1, 2, 3, 4, 5, 6, 7, 8
- Вершина 4 соединена с 1, 3, 4, 5, 6, 7, 8
- Вершина 5 соединена с 1, 2, 3, 4, 5, 6, 7, 8
- Вершина 6 соединена с 1, 2, 3, 4, 5, 6, 7, 8
- Вершина 7 соединена с  2, 3, 4, 5, 6, 7, 8
- Вершина 8 соединена с 1, 2, 3, 4, 5, 6, 7, 8

Для решения этой задачи удобнее всего рассматривать не сам граф $G$, а его **дополнение** $\bar{G}$.
Два подграфа изоморфны в исходном графе тогда и только тогда, когда их дополнения изоморфны в графе-дополнении.

**1. Анализ структуры графа и построение дополнения**
В графе 8 вершин. Посмотрим, какие связи **отсутствуют** (исходя из описания и рисунка):
*   Вершина **1** не соединена с **7** (в списке 1 нет 7, в списке 7 нет 1).
*   Вершина **2** не соединена с **4** (в списке 2 нет 4, в списке 4 нет 2).
*   Вершины **3, 5, 6, 8** соединены со всеми остальными вершинами (являются универсальными).

Таким образом, **граф-дополнение $\bar{G}$** выглядит очень просто:
*   Он содержит всего два ребра: **$(1, 7)$** и **$(2, 4)$**.
*   Остальные вершины ($3, 5, 6, 8$) являются изолированными (в дополнении у них степень 0).

**2. Перебор вариантов выбора 4 вершин**
Нам нужно выбрать любые 4 вершины из 8 и посмотреть, какую фигуру они образуют. Проще смотреть на фигуру, которую они образуют в **дополнении** $\bar{G}$, так как там почти нет ребер.

Возможны всего 3 случая в зависимости от того, сколько целых пар (ребер дополнения) попало в нашу выборку:

*   **Вариант 1: В выборку не попало ни одного полного ребра из $\bar{G}$.**
    *   *Как получить:* Выбрать вершины, которые не соединены в $\bar{G}$. Например, $\{3, 5, 6, 8\}$ (все изолированные) или $\{1, 2, 3, 5\}$ (1 и 2 не соединены в $\bar{G}$).
    *   *Вид в $\bar{G}$:4 изолированные точки* (пустой граф).
    *   *Вид в исходном графе $G$:* **Полный граф $K_4$** (все вершины соединены со всеми).

*   **Вариант 2: В выборку попало ровно одно ребро из $\bar{G}$.**
    *   *Как получить:* Выбрать одну пару (например, 1 и 7) и добавить к ним любые две вершины, не образующие вторую пару (например, 3 и 5). Выборка: $\{1, 7, 3, 5\}$.
    *   *Вид в $\bar{G}$:* 1 ребро и 2 изолированные точки.
    *   *Вид в исходном графе $G$:* Полный граф без одного ребра (**$K_4 - e$**).

*   **Вариант 3: В выборку попали оба ребра из $\bar{G}$.**
    *   *Как получить:* Выбрать вершины $\{1, 7, 2, 4\}$.
    *   *Вид в $\bar{G}$:* 2 изолированных друг от друга ребра.
    *   *Вид в исходном графе $G$:* Полный граф без двух непересекающихся ребер. Это эквивалентно циклу из 4 вершин (**$C_4$**).

Других вариантов быть не может, так как в дополнении всего два ребра. Мы можем взять 0, 1 или 2 из них.

**Ответ:** 3

### Задача 14

1 point possible (graded)

Предложенная разработчиками рекомендательная система некоторой социальной сети довольно агрессивна и предлагает пользователей друг другу, как потенциальных друзей, если у них есть хотя бы один общий друг. Её решили опробовать на небольшой выборке. Рекомендации были разосланы всем одновременно и лишь раз. Известно, что 10 предложений не были одобрены. Сколько образовалось новых пар-друзей?



### Решение:

Для решения этой задачи нужно определить общее количество рекомендаций, которое было сгенерировано системой, и вычесть из него количество отклоненных предложений (10).

**Правило работы системы:** Рекомендация создается для пары пользователей, если они **не являются друзьями**, но имеют **хотя бы одного общего друга**.
Это равносильно поиску всех пар вершин, расстояние между которыми равно 2 (путь вида $A - Друг - B$).

Посчитаем количество таких пар, рассматривая каждого «общего друга» (центральную вершину) и его соседей. Если у вершины $N$ соседей, то она создает $\frac{N(N-1)}{2}$ пар потенциальных рекомендаций между ними.

1.  **Вершина 1** (соседи: 2, 4, 6)
    *   Возможные пары: (2,4), (2,6), (4,6).
    *   Все эти пары не соединены напрямую.
    *   **+3 рекомендации.**

2.  **Вершина 2** (соседи: 1, 5)
    *   Возможная пара: (1,5).
    *   Прямой связи нет.
    *   **+1 рекомендация.**

3.  **Вершина 3** (соседи: 4, 8, 9)
    *   Возможные пары: (4,8), (4,9), (8,9).
    *   Прямых связей нет.
    *   **+3 рекомендации.**

4.  **Вершина 4** (соседи: 1, 3)
    *   Возможная пара: (1,3).
    *   Прямой связи нет.
    *   **+1 рекомендация.**

5.  **Вершина 5** (соседи: 2, 7, 8, 10)
    *   Возможные пары: (2,7), (2,8), (2,10), (7,8), (7,10), (8,10).
    *   Ни одна из этих пар не имеет прямого ребра.
    *   **+6 рекомендаций.**

6.  **Вершина 6** (соседи: 1, 7)
    *   Возможная пара: (1,7).
    *   Прямой связи нет.
    *   **+1 рекомендация.**

7.  **Вершина 7** (соседи: 5, 6)
    *   Возможная пара: (5,6).
    *   Прямой связи нет.
    *   **+1 рекомендация.**

8.  **Вершина 8** (соседи: 3, 5)
    *   Возможная пара: (3,5).
    *   Прямой связи нет.
    *   **+1 рекомендация.**

9.  **Вершины 9 и 10**
    *   Имеют только по 1 соседу, поэтому не могут служить связующим звеном («общим другом») ни для кого.
    *   **0 рекомендаций.**

**Проверка на дубликаты:**
Нужно убедиться, что одну и ту же пару мы не посчитали дважды (это произошло бы, если бы у пары было *два* общих друга, то есть существовал бы цикл из 4 вершин).
При внимательном осмотре графа циклов длины 4 (квадратов) в нем нет. Значит, все найденные пары уникальны.

**Итоговый расчет:**
Общее число рекомендаций = $3 + 1 + 3 + 1 + 6 + 1 + 1 + 1 = 17$.

По условию **10 предложений не были одобрены**.
Значит, количество новых пар (одобренных предложений):
$17 - 10 = 7$

**Ответ:** 7

### Задача 15

Новая версия рекомендательной системы социальной сети уже чуть менее агрессивна и предлагает пользователей друг другу, как потенциальных друзей, если у них есть хотя бы два общих друга. Опять тестируют систему на небольшой выборке. Предположим, рекомендации приходят всем пользователям одновременно раз в день и все пользователи в тот же день одобряют предложение. Как будет выглядеть «граф дружбы» через неделю? Будет ли он ещё меняться? Ниже приведен исходный «граф дружбы».



### Решение:

Для решения задачи проведем пошаговое моделирование процесса добавления друзей («рёбер» графа) день за днем.

**Правило:** Новое соединение (ребро) появляется между двумя вершинами, если у них **нет прямой связи**, но есть **минимум 2 общих соседа**.

### 0. Исходное состояние
Выпишем список соседей для каждой вершины, глядя на граф:
*   **1:** 3, 4, 6
*   **2:** 3, 4, 5
*   **3:** 1, 2, 6, 7
*   **4:** 1, 2
*   **5:** 2, 6
*   **6:** 1, 3, 5
*   **7:** 3

### 1. День первый (Первая волна рекомендаций)
Ищем пары вершин, не соединенные напрямую, у которых 2 или более общих друга.

*   **Пара (1, 2):** Общие друзья — **3, 4**. (2 друга) $\to$ **Добавляем связь (1-2)**.
*   **Пара (2, 6):** Общие друзья — **3, 5**. (2 друга) $\to$ **Добавляем связь (2-6)**.
*   **Пара (3, 4):** Общие друзья — **1, 2**. (2 друга) $\to$ **Добавляем связь (3-4)**.
*   **Пара (3, 5):** Общие друзья — **2, 6**. (2 друга) $\to$ **Добавляем связь (3-5)**.
*   *Остальные пары:*
    *   (1, 5) — общий только 6.
    *   (4, 6) — общий только 1.
    *   (4, 5) — общий только 2.
    *   Вершина 7 имеет только одного друга (3), поэтому она не может иметь $\ge 2$ общих друзей ни с кем.

**Итог первого дня:** Добавились связи: (1-2), (2-6), (3-4), (3-5).


### 2. День второй
Теперь список связей обновился. Проверим, появились ли новые кандидаты.
Вершины 1, 2, 3, 4, 5, 6 стали гораздо сильнее связаны между собой.

*   **Пара (1, 5):** Общие друзья теперь — 2, 3, 6 (раньше был только 6, добавились 2 и 3 благодаря новым связям). $\to$ **Добавляем связь (1-5)**.
*   **Пара (4, 6):** Общие друзья теперь — 1, 2, 3. $\to$ **Добавляем связь (4-6)**.
*   **Пара (4, 5):** Общие друзья теперь — 2, 3. $\to$ **Добавляем связь (4-5)**.

**Что с вершиной 7?**
У вершины 7 по-прежнему только один друг — вершина 3.
Чтобы у 7 появилась связь с кем-то еще (например, с 1), у них должно быть 2 общих друга. Но множество друзей вершины 7 состоит всего из одного элемента $\{3\}$. Пересечение множеств не может быть больше 1.
Следовательно, **вершина 7 больше ни с кем не соединится**.


### 3. День третий и далее
Посмотрим на группу вершин **{1, 2, 3, 4, 5, 6}**.
После второго дня:
*   1 соединена с 2, 3, 4, 5, 6.
*   2 соединена с 1, 3, 4, 5, 6.
*   ... и так далее.
Все вершины с 1 по 6 соединены друг с другом (каждая с каждой). Это называется **полный граф ($K_6$)**.
Внутри этой группы больше добавлять нечего.
Вершина 7 по-прежнему изолирована от всех, кроме 3.

### Ответ
1.  **Как будет выглядеть граф через неделю:**
    Граф стабилизируется уже на второй день. Он будет состоять из двух частей:
    *   **Полный подграф на 6 вершинах** (вершины 1, 2, 3, 4, 5, 6), где "все дружат со всеми".
    *   **Вершина 7**, которая соединена **только с вершиной 3** (как "хвостик", торчащий из полного графа).

2.  **Будет ли он ещё меняться?**
    **Нет**, граф больше меняться не будет. Вершины 1–6 уже имеют максимальное количество связей между собой, а вершина 7 не может набрать 2 общих друзей ни с одним другим участником сети.

### Задача 16

Допустим, нужно связать транспортом города: A, B, C, D, E. Доступно 2 вида связи — автомобиль и самолёт. Пути могут быть только напрямую между парой городов. Стоимость наземного пути будем вычислять по формуле $R S_{\text {д }}+\frac{R S_{\mathrm{a}}}{V_{\mathrm{a}}}$, а воздушного $C+\frac{R S_{\mathrm{c}}}{V_{\mathrm{c}}}$, где R —  расстояние между городами, Sд =1, Sa=300, Va=100, C=1000, Sc=100, Vc=400.

Предложите схему с минимальной суммарной стоимостью при условии того, что из любого города реально будет добраться в любой другой (возможно с пересадками). Какова стоимость такой сети? Ниже приведён граф расстояний между городами.



### Решение

Запишем веса ребер с рисунка : 
AB=150, BC=100, CD=100, DE=400, EA=300, EB=400, EC=400, AC=120, AD=150, BD=150

Для решения задачи нам нужно построить схему с минимальной стоимостью, связывающую все города. Это классическая задача на поиск **Минимального остовного дерева (MST)** в графе.

### Шаг 1. Формулы расчета стоимости
Сначала определим, какой вид транспорта выгоднее использовать для каждого расстояния $R$.

1.  **Наземный путь (Автомобиль):**
    Формула: $Cost_{car} = R \cdot S_д + \frac{R \cdot S_a}{V_a}$
    Подставим значения ($S_д = 1, S_a = 300, V_a = 100$):
    $$Cost_{car} = R \cdot 1 + \frac{R \cdot 300}{100} = R + 3R = \mathbf{4R}$$

2.  **Воздушный путь (Самолёт):**
    Формула: $Cost_{plane} = C + \frac{R \cdot S_c}{V_c}$
    Подставим значения ($C = 1000, S_c = 100, V_c = 400$):
    $$Cost_{plane} = 1000 + \frac{R \cdot 100}{400} = \mathbf{1000 + 0.25R}$$

**Точка перехода:**
Выясним, при каком расстоянии самолет становится выгоднее машины:
$1000 + 0.25R < 4R$
$1000 < 3.75R$
$R > 266.6$
Следовательно:
*   Для расстояний $\le 266$ выгоднее **машина**.
*   Для расстояний $> 266$ выгоднее **самолет**.

### Шаг 2. Расчет стоимости ребер (весов графа)
Рассчитаем оптимальную стоимость проезда для каждого ребра, выбирая более дешевый транспорт.

**Группа 1: Короткие расстояния (используем машину, $4R$)**
*   **$BC$** ($R=100$): $100 \cdot 4 = \mathbf{400}$
*   **$CD$** ($R=100$): $100 \cdot 4 = \mathbf{400}$
*   **$AC$** ($R=120$): $120 \cdot 4 = \mathbf{480}$
*   **$AB$**, **$AD$**, **$BD$** ($R=150$): $150 \cdot 4 = \mathbf{600}$

**Группа 2: Длинные расстояния (используем самолет, $1000 + 0.25R$)**
*   **$EA$** ($R=300$): $1000 + 0.25 \cdot 300 = 1000 + 75 = \mathbf{1075}$
    *(Для сравнения, на машине было бы $300 \cdot 4 = 1200$)*
*   **$DE$**, **$EB$**, **$EC$** ($R=400$): $1000 + 0.25 \cdot 400 = 1000 + 100 = \mathbf{1100}$

### Шаг 3. Построение минимального остовного дерева
Теперь выберем самые дешевые ребра, чтобы соединить все 5 городов ($A, B, C, D, E$), избегая циклов (алгоритм Краскала). Нам нужно выбрать 4 ребра.

1.  Самое дешевое ребро: **$BC$** (стоимость **400**). Соединяем $B-C$.
2.  Следующее дешевое: **$CD$** (стоимость **400**). Соединяем $C-D$. Теперь у нас цепочка $B-C-D$.
3.  Следующее дешевое: **$AC$** (стоимость **480**). Соединяем $A$ с группой. Теперь $A-C-D-B$.
    *(Ребра $AB, AD, BD$ пропускаем, так как они стоят 600, а все эти вершины уже связаны более дешевыми путями).*
4.  Осталось присоединить город **$E$**. Самое дешевое ребро, ведущее к $E$ — это **$EA$** (стоимость **1075**).
    *(Остальные ребра к E стоят 1100).*

### Итоговая схема и стоимость
Схема маршрутов:
1.  **$B \leftrightarrow C$** (Автомобиль): 400
2.  **$C \leftrightarrow D$** (Автомобиль): 400
3.  **$A \leftrightarrow C$** (Автомобиль): 480
4.  **$E \leftrightarrow A$** (Самолёт): 1075

**Суммарная стоимость:**
$$400 + 400 + 480 + 1075 = \mathbf{2355}$$

**Ответ:** 2355

### Задача 17

Исследуем некую социальную сеть. Назовём ключевыми пользователями тех, при удалении которых граф распадётся на максимальное количество компонент связности. Найдите всех ключевых пользователей для этой небольшой выборки:



### Решение

Список связей (смежности) по изображению:
0 соединен с: 1, 9
1 соединен с: 0, 8
2 соединен с: 5, 6, 7
3 соединен с: 7, 8
4 соединен с: 7
5 соединен с: 2, 9
6 соединен с: 2
7 соединен с: 2, 3, 4
8 соединен с: 1, 3, 9
9 соединен с: 0, 5, 8


Проанализируем граф с учетом этих данных.

**Цель:** Найти вершины ("ключевых пользователей"), при удалении которых граф распадается на **максимальное** количество частей (компонент связности).

**1. Анализ "листьев" (висячих вершин)**
Посмотрим на вершины, у которых всего **одна связь**:
*   **Вершина 6**: соединена только с **2**.
*   **Вершина 4**: соединена только с **7**.

**2. Проверка удаления вершин, связанных с листьями**
Если удалить вершину, к которой прикреплен "лист", этот "лист" отвалится и станет изолированным островком.

*   **Удаляем вершину 2:**
    *   **Вершина 6** теряет единственную связь и становится изолированной (Компонента №1).
    *   Остальной граф: Вершина 5 (которая была связана с 2) не теряется, так как у нее есть запасная связь с 9. Вершина 7 (которая была связана с 2) имеет запасные связи с 3 и 4. Остальная сеть остается связной благодаря циклу $0-1-8-9-5...$ и другим путям.
    *   **Итог:** Граф распадается на **2 компоненты**.

*   **Удаляем вершину 7:**

### Задача 21

Транспортная сеть некоторого региона устроена следующим образом: есть n городов, и любые два из них имеют либо прямое наземное сообщение, либо прямое воздушное (но не одновременно). Известно, что схемы наземного и воздушного сообщения подобны (то есть изоморфны как графы). Докажите, что либо n, либо n−1 делится на 4.

### Решение:

Для решения этой задачи воспользуемся свойствами графов и элементарной теорией чисел.

**1. Формализация задачи на языке теории графов**

Представим города как вершины графа, а маршруты — как рёбра.
*   Пусть $n$ — количество городов (вершин).
*   По условию, **любые два** города соединены ровно одним типом связи. Это означает, что если объединить все наземные и все воздушные маршруты, мы получим **полный граф** $K_n$ (граф, в котором каждая вершина соединена с каждой).
*   Пусть $G$ — граф наземного сообщения, а $\bar{G}$ — граф воздушного сообщения (дополнение к графу $G$).
*   По условию, схемы наземного и воздушного сообщения **изоморфны**, то есть $G \cong \bar{G}$. Такие графы называются **самодополнительными**.

**2. Подсчет количества рёбер**

В полном графе $K_n$ общее количество рёбер вычисляется по формуле:
$$|E| = \frac{n(n - 1)}{2}$$

Так как графы наземного ($G$) и воздушного ($\bar{G}$) сообщения изоморфны, количество рёбер в них должно быть одинаковым:
$$|E(G)| = |E(\bar{G})|$$

Поскольку эти два графа в сумме составляют полный граф, сумма их рёбер равна общему числу рёбер в $K_n$:
$$|E(G)| + |E(\bar{G})| = \frac{n(n - 1)}{2}$$

Заменим $|E(\bar{G})|$ на $|E(G)|$:
$$2 \cdot |E(G)| = \frac{n(n - 1)}{2}$$

Отсюда выразим количество рёбер в наземном графе:
$$|E(G)| = \frac{n(n - 1)}{4}$$

**3. Анализ условия делимости**

Количество рёбер $|E(G)|$ обязательно должно быть **целым числом**.
Это значит, что выражение $\frac{n(n - 1)}{4}$ должно быть целым.
Следовательно, произведение $n(n - 1)$ должно делиться на 4 без остатка:
$$n(n - 1) \equiv 0 \pmod 4$$

**4. Рассмотрение возможных вариантов для $n$**

Число $n$ и число $(n-1)$ — это два последовательных целых числа. Одно из них обязательно чётное, а другое — нечётное.
Произведение чётного и нечётного чисел делится на 4 тогда и только тогда, когда **чётный множитель сам делится на 4** (так как нечётный множитель не содержит двоек в разложении, а для деления на 4 нужны две двойки).

Рассмотрим два случая:

1.  **Если $n$ — чётное:**
    Тогда $(n-1)$ — нечётное. Чтобы произведение $n(n-1)$ делилось на 4, число $n$ обязано делиться на 4.
    $$\Rightarrow n \text{ делится на } 4.$$

2.  **Если $n$ — нечётное:**
    Тогда $(n-1)$ — чётное. Чтобы произведение $n(n-1)$ делилось на 4, число $(n-1)$ обязано делиться на 4.
    $$\Rightarrow (n - 1) \text{ делится на } 4.$$



**Вывод**

Мы доказали, что условие существования самодополнительного графа выполняется только тогда, когда произведение $n(n-1)$ кратно 4. Это возможно только в двух случаях: либо само $n$ кратно 4, либо $(n-1)$ кратно 4.
