### Задание 1.9

1 point possible (graded)

Была построена модель машинного обучения, которая решает задачу классификации файлов на безопасные (класс 0) и вирусные (класс 1). После валидации модели на 100 наблюдениях была получена следующая матрица ошибок:
```
        y^=0	y^=1
y=0	    73	    13
y=1	    4	    10
```
Здесь y — истинные классы файлов в валидационной выборке, а y^ — предсказания модели.

Рассчитайте условную вероятность того, что файл является вирусным, при условии, что модель посчитала файл вирусным: P(y=1| y^=1). Иначе говоря, рассчитайте метрику precision для обученной модели. Ответ округлите до двух знаков после точки-разделителя.

Рассчитаем эту условную вероятность, используя данные из матрицы ошибок.

Матрица ошибок выглядит следующим образом:

```
                 y^=0  y^=1 (Предсказано)
y=0  (Истина)      73    13   (Фактически безопасные)
y=1                 4    10   (Фактически вирусные)
```

* **y=0**: Истинно безопасный файл
* **y=1**: Истинно вирусный файл
* **y^=0**: Модель предсказала как безопасный
* **y^=1**: Модель предсказала как вирусный

Нам нужно рассчитать условную вероятность того, что файл является вирусным ($y=1$), при условии, что модель посчитала файл вирусным ($\hat{y}=1$). Это $P(y=1 | \hat{y}=1)$.

По определению условной вероятности:
$P(y=1 | \hat{y}=1) = \frac{P(y=1 \cap \hat{y}=1)}{P(\hat{y}=1)}$

* $P(y=1 \cap \hat{y}=1)$ - это вероятность того, что файл *одновременно* является вирусным (y=1) и был предсказан как вирусный ($\hat{y}=1$). Количество таких файлов в матрице ошибок равно 10 (истинно положительные, TP).
* $P(\hat{y}=1)$ - это вероятность того, что модель предсказала файл как вирусный ($\hat{y}=1$). Количество таких файлов - это сумма тех, которые были предсказаны как вирусные: 13 (ложноположительные, FP) + 10 (истинно положительные, TP) = 23.

В терминах количества наблюдений из матрицы ошибок, условная вероятность $P(y=1 | \hat{y}=1)$ рассчитывается как:

$P(y=1 | \hat{y}=1) = \frac{\text{Количество истинно положительных (TP)}}{\text{Количество всех предсказаний класса 1 (TP + FP)}}$

В данном случае:
* TP = 10
* FP = 13

Следовательно:
$P(y=1 | \hat{y}=1) = \frac{10}{10 + 13} = \frac{10}{23}$

Это и есть метрика Precision для класса 1 (вирусный).

Рассчитаем значение и округлим его до двух знаков после точки:
$\frac{10}{23} \approx 0.43478...$

Округляем до 0.43.

Вот код на Python для выполнения этого расчета:

```python

**Результат выполнения кода:**

```
Истинно положительные (TP): 10
Ложноположительные (FP): 13
Всего предсказаний класса 1 (TP + FP): 23
Рассчитанное Precision (P(y=1 | y^=1)): 0.43478260869565216
Precision, округленное до двух знаков: 0.43
```

**Ответ:** Условная вероятность того, что файл является вирусным, при условии, что модель посчитала файл вирусным, составляет **0.43**.

````python
# Данные из матрицы ошибок
# Строки - истинные классы (y)
# Столбцы - предсказанные классы (y^)

# y=0 (Истина: безопасный)
true_negative = 73  # y=0, y^=0
false_positive = 13 # y=0, y^=1

# y=1 (Истина: вирусный)
false_negative = 4  # y=1, y^=0
true_positive = 10  # y=1, y^=1

# Расчет Precision для класса 1 (вирусный)
# Precision = TP / (TP + FP)
# Это вероятность P(y=1 | y^=1)
precision = true_positive / (true_positive + false_positive)

# Округление до двух знаков после точки
rounded_precision = round(precision, 2)

print(f"Истинно положительные (TP): {true_positive}")
print(f"Ложноположительные (FP): {false_positive}")
print(f"Всего предсказаний класса 1 (TP + FP): {true_positive + false_positive}")
print(f"Рассчитанное Precision (P(y=1 | y^=1)): {precision}")
print(f"Precision, округленное до двух знаков: {rounded_precision}")

````

Найдите угол между векторами  (1;0;3)  и  (−6;7;2) .

````python
a = np.array([1, 0, 3])
b = np.array([-6, 7, 2])

alpha = np.dot(a, b) / (np.linalg.norm(a) * np.linalg.norm(b))
angle = np.arccos(alpha)
print(f"Угол между векторами: {angle} радиан")
print(f"Угол между векторами: {angle*180/np.pi} градусов")
````

Посчитайте межквартильный размах для ряда чисел  8,23,20,18,20,16,6 .

````python
data = np.array([8,23,20,18,20,16,6])
# Сортируем данные
sorted_data = np.sort(data)

# Находим медиану
median = np.median(sorted_data) 

# Делим выборку на две части
lower_half = sorted_data[:len(sorted_data)//2]
upper_half = sorted_data[len(sorted_data)//2:]

# Находим медианы обеих половин
lower_median = np.median(lower_half)
upper_median = np.median(upper_half)

# Вычисляем межквартильный размах
iqr = upper_median - lower_median

print(f"Межквартильный размах для ряда ({', '.join(map(str, data))}): {iqr}")

````
