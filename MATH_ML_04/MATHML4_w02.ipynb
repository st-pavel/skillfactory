```python
from sympy import Symbol
from helper.functionanalyzer import FunctionAnalyzer

x = Symbol('x')

# Пример 1: функция с наклонной асимптотой y = x + 1
f1 = (x**2 + x)/(x - 1)
analyzer1 = FunctionAnalyzer(f1, x)
analyzer1.describe()  # Покажет наклонную асимптоту: y = x + 1
analyzer1.plot()

# Пример 2: функция с наклонной асимптотой y = 2x - 3
f2 = 2*x + (x-3)/(x+1)
analyzer2 = FunctionAnalyzer(f2, x)
analyzer2.describe()  # Покажет наклонную асимптоту: y = 2x - 3
analyzer2.plot()

# Пример 3: функция с наклонной асимптотой y = -x + 2
f3 = -x + 2 + 1/x
analyzer3 = FunctionAnalyzer(f3, x)
analyzer3.describe()  # Покажет наклонную асимптоту: y = -x + 2
analyzer3.plot()
```
