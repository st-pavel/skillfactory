```python
# Готовое решение
import pandas as pd
import numpy as np


df = pd.read_csv('test_data.csv', index_col=0)
#df = pd.read_csv('test_data.csv') # в решении - оставил эту строку, чтобы не было ошибки. Верхнюю строку закомментировал, т.к. неправильно импортировал данные индекса

#print(df.head(10))
# Шаг 1: Удаление признаков с более чем 50% пропущенных значений
threshold = len(df) * 0.5
df = df.dropna(thresh=threshold, axis=1)

# Шаг 2: Удаление строк с более чем двумя пропусками
df = df.dropna(thresh=len(df.columns) - 2)

# Шаг 3: Заполнение числовых признаков средним значением
numeric_cols = df.select_dtypes(include=[np.number]).columns
df[numeric_cols] = df[numeric_cols].fillna(df[numeric_cols].mean())

# Заполнение категориальных признаков модой
categorical_cols = df.select_dtypes(include=[object]).columns
for col in categorical_cols:
    df[col] = df[col].fillna(df[col].mode()[0])

print(df.head(10))
```
