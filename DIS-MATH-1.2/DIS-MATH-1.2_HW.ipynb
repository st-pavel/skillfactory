### Задача 3.8

1/1 point (graded)

На собеседованиях часто встречаются задачи вида «посчитайте количество определённых чисел» или задачи на шахматы. Эти задачи решаются методами комбинаторики. Обычно задача сформулирована в виде «Сколько существует чисел ...» или «Сколько существует расстановок шахмат ...», а потом идёт некоторое условие.

Сколько существует четырехзначных чисел состоящих только из цифр 1, 2, 3, 4?


Это задача на размещение с повторениями.

### Ход решения

Представляем четырехзначное число как 4 пустые ячейки:
`[ ] [ ] [ ] [ ]`

У нас есть набор допустимых цифр: **{1, 2, 3, 4}**. Всего их **4 штуки**.

1.  **Первая цифра:** Мы можем поставить любую из 4-х цифр. (4 варианта)
2.  **Вторая цифра:** Мы снова можем поставить любую из 4-х цифр (повторения разрешены, условие не запрещает число 1111). (4 варианта)
3.  **Третья цифра:** Аналогично, любая из 4-х. (4 варианта)
4.  **Четвертая цифра:** Любая из 4-х. (4 варианта)

По правилу произведения перемножаем количество вариантов:
$$4 \times 4 \times 4 \times 4 = 4^4 = \textbf{256}$$


Ответ: **256**.

### Задача 3.7
1/1 point (graded)

Компания выпускает 5 видов программного обеспечения. Менеджмент задумался о том, чтобы расширить линейку продаж за счёт того, чтобы продавать объединённые пакеты ПО, например, первый и второй вид вместе или все пять видов сразу.

Сколько дополнительных продуктов формируется при таком пополнении?

### Решение

Здесь мы работаем с формулой $2^n$, но с важным уточнением: нам нужно исключить "одиночные" товары и "пустой" набор.

**1. Считаем вообще все возможные комбинации**
У нас 5 видов ПО. Каждый вид мы можем либо включить в пакет, либо нет.
Всего вариантов наборов:
$$2^5 = 32$$

**2. Исключаем "пустой" пакет**
Один из этих 32 вариантов — это когда мы не выбрали ни одной программы (00000). Продавать воздух мы не будем.
$$32 - 1 = 31 \text{ вариант продажи.}$$
*(Кстати, если бы вопрос звучал "сколько всего вариантов покупки существует", ответ был бы 31).*

**3. Исключаем уже существующие товары**
Вопрос звучит: «Сколько **дополнительных** продуктов формируется за счёт того, чтобы продавать **объединённые** пакеты».
Обычные одиночные программы (Программа 1, Программа 2... Программа 5) у компании уже есть. Это не *объединённые* пакеты. Таких вариантов ровно 5.

Вычитаем их из общего числа:
$$31 - 5 = 26$$

---

### Альтернативный способ (через сложение)
Можно посчитать только пакеты, состоящие из 2 и более программ, используя треугольник Паскаля (строка 5: `1 5 10 10 5 1`):

* Пакеты из **2** программ ($C_5^2$): **10** штук
* Пакеты из **3** программ ($C_5^3$): **10** штук
* Пакеты из **4** программ ($C_5^4$): **5** штук
* Пакет из всех **5** программ ($C_5^5$): **1** штука

Складываем:
$$10 + 10 + 5 + 1 = \textbf{26}$$

### Резюме
Используем формулу:
$$N = 2^n - n - 1$$
Где $2^n$ — все подмножества, $n$ — одиночные элементы, $1$ — пустое множество.


Ответ: **26 дополнительных продуктов**.

### Задача 4.8

1/1 point (graded)

Число в компьютере записывается последовательностью из 32 элементов, каждый из которых или 0, или 1. Первое число обозначает знак числа: если 0, то число отрицательное, если 1, то число положительное. Остальные элементы последовательности кодируют модуль числа в двоичной записи.

Сколько чисел можно закодировать таким образом?


### Решение.


В этой задаче есть «ловушка», связанная с тем, как мы считаем ноль. Давайте разберем решение поэтапно.

### 1. Общее количество комбинаций битов
У нас есть 32 ячейки (бита), каждая из которых может быть 0 или 1.
Общее количество возможных комбинаций из нулей и единиц вычисляется по формуле $2^N$:
$$2^{32} = 4\ 294\ 967\ 296$$

Если бы вопрос звучал «Сколько различных **кодов** можно составить», мы бы остановились на этом числе.

### 2. Проблема «Двойного нуля»
В условии сказано, что используется **прямой код** (Sign-Magnitude):
* 1 бит — знак (+ или -).
* 31 бит — модуль (значение).

Давайте посмотрим, что происходит, когда модуль числа равен нулю (все 31 бит равны 0):
1.  **Знак 1 (Положительный) + Модуль 0** $\rightarrow$ Получаем **+0**.
2.  **Знак 0 (Отрицательный) + Модуль 0** $\rightarrow$ Получаем **-0**.

С точки зрения математики, $+0$ и $-0$ — это **одно и то же число** (просто ноль).
Компьютер в такой системе кодирования хранит их как две разные последовательности битов, но математическое значение у них одно.

### 3. Итоговый расчет
Поскольку вопрос звучит «Сколько **чисел** можно закодировать таким образом», мы должны исключить дубликат нуля.

$$N = 2^{32} - 1 = 4\ 294\ 967\ 295$$


### Проверка решения по структуре данных
Как распределяются эти числа:
* **Положительные числа:** $2^{31} - 1$ (от 1 до максимума).
* **Отрицательные числа:** $2^{31} - 1$ (от -1 до минимума).
* **Ноль:** 1 число (которое занимает два кода).

$$(2^{31} - 1) + (2^{31} - 1) + 1 = 2^{32} - 1$$


Ответ: $$ 4\ 294\ 967\ 295$$

### Задача 4.4

1/1 point (graded)

Сколько существует способов расставить 2 королей (чёрного и белого) на шахматной доске так, чтобы они не били друг друга?


### Решение

Король «бьет» все соседние клетки (по диагонали, вертикали и горизонтали). Количество запретных клеток для второго короля зависит от того, где стоит первый.

В зависимости от положения первого кородя, мы разбиваем шахматную доску на три зоны: Углы, Края (без углов) и Центр.

Используем правило суммы вариантов.

### Шаг 1: Если первый король (например, Белый) стоит в углу
* **Количество углов:** 4 клетки (a1, a8, h1, h8).
* **Зона поражения:** Король в углу бьет 3 клетки + занимает 1 сам. Итого 4 клетки недоступны.
* **Свободных мест для Черного короля:** $64 - 4 = 60$.

$$4 \times 60 = 240 \text{ способов}$$

### Шаг 2: Если первый король стоит на краю доски (не в углу)
* **Количество таких клеток:** Периметр доски минус углы.
    * Всего по краям: $8 + 8 + 6 + 6 = 28$ клеток.
    * Убираем 4 угла: $28 - 4 = 24$ клетки.
* **Зона поражения:** Король на краю бьет 5 клеток + занимает 1 сам. Итого 6 клеток недоступны.
* **Свободных мест для Черного короля:** $64 - 6 = 58$.

$$24 \times 58 = 1392 \text{ способа}$$

### Шаг 3: Если первый король стоит в центре (не на краю)
* **Количество таких клеток:** Это квадрат $6 \times 6$ внутри доски.
    * $6 \times 6 = 36$ клеток.
* **Зона поражения:** Король в центре бьет 8 клеток вокруг себя + занимает 1 сам. Итого 9 клеток недоступны.
* **Свободных мест для Черного короля:** $64 - 9 = 55$.

$$36 \times 55 = 1980 \text{ способов}$$



### Итоговый расчет
Складываем все варианты:
$$240 + 1392 + 1980 = \textbf{3 612}$$

**Примечание:** Так как короли разного цвета (белый и чёрный), ситуация «Белый на a1, Чёрный на h8» и «Чёрный на a1, Белый на h8» — это разные расстановки. Вышеизложенный метод подсчета это учитывает автоматически.



Ответ: **3 612 способов**.

1.  Перед первым чёрным.
2.  Между первым и вторым.
3.  Между вторым и третьим.
4.  После третьего.

Нам нужно выбрать **2 места** из этих **4** для размещения двух белых шаров. Порядок не важен (шары одинаковые).
Используем формулу сочетаний:
$$N_{Merab} = C_4^2 = \frac{4 \times 3}{2} = 6$$

**Варианты Мераба (для проверки):**
1.  Места (1, 2) $\rightarrow$ **Б Ч Б** Ч Ч
2.  Места (1, 3) $\rightarrow$ **Б** Ч Ч **Б** Ч
3.  Места (1, 4) $\rightarrow$ **Б** Ч Ч Ч **Б**
4.  Места (2, 3) $\rightarrow$ Ч **Б Ч Б** Ч
5.  Места (2, 4) $\rightarrow$ Ч **Б** Ч Ч **Б**
6.  Места (3, 4) $\rightarrow$ Ч Ч **Б Ч Б**

**Итог Мераба:** **6 вариантов**.

10. **Ч Ч Ч Б Б** Осип


#### Ответ

При такой трактовке условий:

### Решение 3

Рассмотрим Вариант 3 трактовки условий задания:

Для Осипа выигрышем будет комбинация ЧББ **не подряд** (т.е. например ЧБЧБ), а для Мераба выигрышная комбинация -  **"чёрный шар вышел между белыми"**, Это значит, что белые шары **не стоят рядом** (т.е. для Мераба выигрышной ситуацией будет, например БЧЧБ ).


#### 1. Аналитическое решение для Осипа

**Условие:** *«После какого-то из чёрных шаров вышли два белых»* (не обязательно подряд).
То есть, в последовательности должен быть хотя бы один Чёрный шар, который стоит левее обоих Белых шаров.

**Логика:**
Давай посмотрим, когда Осип **проигрывает**.
Чтобы условие *не выполнилось*, ни один чёрный шар не должен стоять перед парой белых.
Это возможно только в одной ситуации: если **первый же шар — белый**.

* Если первый шар **Белый** ($W_1...$), то в мешке остался только один белый шар ($W_2$). Даже если потом пойдут чёрные шары, после них будет максимум один белый. Условие «два белых после чёрного» выполнить невозможно физически.
* Если первый шар **Чёрный** ($B_1...$), то в мешке остались оба белых шара. Значит, как бы они ни расположились дальше, они гарантированно будут стоять *после* этого первого чёрного шара.

**Расчет:**
Осип выигрывает тогда и только тогда, когда на первой позиции стоит Чёрный шар.
У нас 5 позиций. Чёрных шаров 3.
Вероятность того, что на первой позиции чёрный:
$$P = \frac{3}{5}$$

Количество выигрышных вариантов:
$$10 \times \frac{3}{5} = \textbf{6 вариантов}.$$


#### 2. Аналитическое решение для Мераба

**Условие:** *«Чёрный шар вышел между белыми»* (Белые не стоят рядом).
То есть запрещена комбинация $ББ$.

**Логика (Метод слотов/перегородок):**
Чтобы белые шары гарантированно не стояли рядом, мы сначала выставляем в ряд все чёрные шары, создавая между ними "слоты" для белых.

Схема чёрных шаров:
`_ Ч _ Ч _ Ч _`

У нас есть **4** возможных места (слота), куда можно положить белые шары:
1.  До первого чёрного.
2.  Между 1-м и 2-м.
3.  Между 2-м и 3-м.
4.  После последнего.

Чтобы белые шары не соприкасались, нам нужно положить их в **два разных слота** из 4-х доступных.

**Расчет:**
Число способов выбрать 2 слота из 4-х:
$$C_4^2 = \frac{4 \times 3}{2} = \textbf{6 вариантов}.$$

Выпишем варианты все (Б = Белый, Ч = Чёрный):

1.  **Б Б Ч Ч Ч** Ничья
2.  **Б Ч Б Ч Ч** Мераб
3.  **Б Ч Ч Б Ч** Мераб
4.  **Б Ч Ч Ч Б** Мераб
5.  **Ч Б Б Ч Ч** Осип
6.  **Ч Б Ч Б Ч** Осип и Мераб
7.  **Ч Б Ч Ч Б** Осип и Мераб
8.  **Ч Ч Б Б Ч** Осип
9.  **Ч Ч Б Ч Б** Осип и Мераб
10. **Ч Ч Ч Б Б** Осип


#### Итоговый вывод

* **Осип:** 6 выигрышных вариантов.
* **Мераб:** 6 выигрышных вариантов.

**Ответ:** При такой трактовке условий (Осип — «мягкое» условие, Мераб — «отсутствие ББ») **шансы игроков абсолютно равны**.

Это единственный сценарий, где игра становится справедливой. 

Любопытно, что при этом их условия не исключают друг друга:
Например, комбинация **Ч Б Ч Б Ч** является выигрышной **для обоих**:
* Первый шар Чёрный $\rightarrow$ Осип выиграл.
* Белые разделены $\rightarrow$ Мераб выиграл.



### Решение 4

Рассмотрим Вариант 4 трактовки условий задания:

Для Осипа выигрышем будет комбинация ЧББ **не подряд** (т.е. два белых шара следуют после черного, например ЧБЧЧБ), а для Мераба выигрышная комбинация - БЧБ **подряд** (строго).

Комбинация условий: «мягкое» (вероятностное) условие для одного и «жесткое» (структурное) для другого.

#### 1. Аналитическое решение для Осипа

**Условие:** *«После какого-то из чёрных шаров вышли два белых»* (не обязательно подряд).
Пример: `Ч ... Б ... Б`.

**Логика решения (через первый шар):**
Чтобы условие выполнилось, необходимо наличие хотя бы одного чёрного шара **раньше**, чем закончится набор белых шаров. А точнее, раньше, чем появится *первый* белый? Нет, условие требует, чтобы *оба* белых были после черного.

Это означает, что **самый первый шар в последовательности обязан быть чёрным**.
* Если первый шар **Чёрный** ($Ч_1...$), то за ним следуют ещё 4 шара, среди которых находятся оба белых. Значит, оба белых шара гарантированно находятся *после* первого чёрного. Условие выполнено.
* Если первый шар **Белый** ($Б_1...$), то этот белый шар уже выпал *до* любого чёрного. Условие «два белых после чёрного» выполнить невозможно (так как один уже выскочил раньше).

**Расчет:**
Вероятность того, что на первой позиции стоит чёрный шар:
$$P = \frac{\text{Количество чёрных}}{\text{Всего шаров}} = \frac{3}{5}$$

Количество выигрышных вариантов:
$$10 \times \frac{3}{5} = \textbf{6 вариантов}.$$


#### 2. Аналитическое решение для Мераба

**Условие:** *«Белый, Чёрный, Белый подряд»* (блок **БЧБ**).

**Логика решения (через склеивание):**
Мы ищем строго фиксированный паттерн. Представим последовательность `[БЧБ]` как один неделимый супер-объект.

Для составления этого супер-объекта мы расходуем:
* 2 белых шара (все, что были).
* 1 чёрный шар.

**Что осталось расставить:**
1.  Наш супер-объект `[БЧБ]`.
2.  Оставшийся чёрный шар `Ч`.
3.  Ещё один оставшийся чёрный шар `Ч`.

Итого у нас **3 объекта**, которые нужно расставить по местам. Из них два объекта одинаковые (`Ч`), а один уникальный (`[БЧБ]`).

**Расчет:**
Число перестановок с повторениями:
$$N = \frac{3!}{2! \cdot 1!} = \frac{3 \times 2 \times 1}{2} = \textbf{3 варианта}.$$

**Проверка (визуализация этих 3 вариантов):**
1.  `[БЧБ] Ч Ч` (Паттерн в начале)
2.  `Ч [БЧБ] Ч` (Паттерн в центре)
3.  `Ч Ч [БЧБ]` (Паттерн в конце)


Выпишем варианты все (Б = Белый, Ч = Чёрный):

1.  **Б Б Ч Ч Ч** Ничья
2.  **Б Ч Б Ч Ч** Мераб
3.  **Б Ч Ч Б Ч** Ничья
4.  **Б Ч Ч Ч Б** Ничья
5.  **Ч Б Б Ч Ч** Осип
6.  **Ч Б Ч Б Ч** Осип и Мераб
7.  **Ч Б Ч Ч Б** Осип
8.  **Ч Ч Б Б Ч** Осип
9.  **Ч Ч Б Ч Б** Осип и Мераб
10. **Ч Ч Ч Б Б** Осип

#### Итоговый вывод для Варианта 4

* **Осип:** 6 выигрышных комбинаций.
* **Мераб:** 3 выигрышные комбинации.


Статистика по списку:

Всего выигрышей Осипа: № 5, 6, 7, 8, 9, 10 = 6 раз (что совпадает с нашим аналитическим расчетом $10 \times \frac{3}{5}$).
Всего выигрышей Мераба: № 2, 6, 9 = 3 раза (что совпадает с расчетом перестановок блока БЧБ).

Ничья (никто не получил): № 1, 3, 4 = 3 раза.

Интересно - в этой трактовке условий есть случаи одновременного выигрыша (№6 и №9), которые часто упускают из виду при ручном подсчете.

**Ответ:** При такой постановке (Осипу — "мягкое" условие, Мерабу — "жесткий" блок) игра **выгоднее Осипу ровно в 2 раза**.
