```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px
```

```python
df = pd.read_csv('new-site.csv', sep="\t")

```

```python
# Группируем данные по колонке 'site' и суммируем значения в колонке 'dwell-time'
total_dwell_time = df.groupby('site')['dwell-time'].sum()

# Выводим результат
print(total_dwell_time)
```

```python
# Группируем данные по колонке 'site' и вычисляем среднее значение в колонке 'dwell-time'
mean_dwell_time = df.groupby('site')['dwell-time'].mean().round()

# Выводим результат
print(mean_dwell_time)
```

```python
from scipy.stats import shapiro

# Проверка на нормальность для группы А (site = 0)
group_A = df[df['site'] == 0]['dwell-time']
stat_A, p_value_A = shapiro(group_A)
print(f"Группа А: статистика = {stat_A}, p-значение = {p_value_A}")

# Проверка на нормальность для группы В (site = 1)
group_B = df[df['site'] == 1]['dwell-time']
stat_B, p_value_B = shapiro(group_B)
print(f"Группа В: статистика = {stat_B}, p-значение = {p_value_B}")

# Интерпретация результатов
alpha = 0.05
if p_value_A > alpha:
    print("Группа А: Распределение нормальное (не отвергаем H₀)")
else:
    print("Группа А: Распределение не нормальное (отвергаем H₀)")

if p_value_B > alpha:
    print("Группа В: Распределение нормальное (не отвергаем H₀)")
else:
    print("Группа В: Распределение не нормальное (отвергаем H₀)")
```

```python
from scipy.stats import mannwhitneyu

# Разделяем данные на группы
group_A = df[df['site'] == 0]['dwell-time']
group_B = df[df['site'] == 1]['dwell-time']

# Выполняем тест Манна — Уитни
stat, p_value = mannwhitneyu(group_A, group_B, alternative='two-sided')

# Выводим результаты
print(f"Статистика теста: {stat}")
print(f"p-значение: {p_value}")

# Интерпретация результатов
alpha = 0.05
if p_value > alpha:
    print("Нет оснований отвергать H₀: Время пребывания в группах А и В совпадает.")
else:
    print("Отвергаем H₀: Время пребывания в группах А и В различается.")
```
