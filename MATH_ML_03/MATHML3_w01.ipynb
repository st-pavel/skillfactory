Найдите собственные числа и собственные вектора матрицы:

A = 3, 2
    1, 2


```python
# создаем матрицу А
A = np.array([
    [3, 1],
    [2, 2]
]).T
# вычисляем собственные числа и собственные векторы
eig_values, eig_vectors = np.linalg.eig(A)
print('Собственные числа: \n', eig_values)
print('Собственные векторы: \n', eig_vectors)
```

Тут сразу возникают вопросы. Полученные собственные векторы вовсе не совпадают с тем, что мы получали в ручном счёте. Секрет в том, что в numpy при расчёте собственных векторов они автоматически нормализуются (как векторы) — приводятся к длине, равной 1. Если вы стандартизируете вектора, которые мы получали в теоретической части, вы получите ровно те же самые вектора. Можете проверить самостоятельно. Напомним, что стандартизацию векторов мы проходили в модуле MATH&ML-2. Линейная алгебра в контексте линейных методов ч. 2.

Найдите собственные числа матрицы:
     1   4   5
A = -5   8   9
    -6   7   11

```python
# создаем матрицу А
A = np.array([
    [1, 4, 5],
    [-5, 8, 9 ],
    [-6, 7, 11]
]).T
# вычисляем собственные числа и собственные векторы
eig_values, eig_vectors = np.linalg.eig(A)
print('Собственные числа: \n', eig_values)
print('Собственные векторы: \n', eig_vectors)
```

Найдём собственные числа заведомо вырожденной матрицы. 

      1  4   5   8
     -4 12  14  15
A =  -5  8   9   7
     -6  7   11  4 

```python
# создаем матрицу А
A = np.array([
    [1, 4, 5, 8],
    [-4, 12, 14, 15],
    [-5, 8, 9, 7],
    [-6, 7, 11, 4]
]).T
# вычисляем собственные числа и собственные векторы
eig_values, eig_vectors = np.linalg.eig(A)
print('Собственные числа: \n', eig_values)
print('Собственные векторы: \n', eig_vectors)
```

Что видим? Одно из собственных чисел получилось равным 0. Это сигнал к тому, что строки или столбцы матрицы A являются линейно зависимыми, то есть матрица плохо обусловлена, а точнее — вырождена.

Задание 4.2
4 points possible (ungraded)
Найдите собственные числа матрицы Грама ATA, если:

     1    9    4     
A =  9    4    7
     4    7    12

```python
# создаем матрицу А
A = np.array([
    [1, 9, 4],
    [9, 4, 7],
    [4, 7, 12]
]).T
# вычисляем собственные числа и собственные векторы
eig_values, eig_vectors = np.linalg.eig(A.T @ A)
print('Собственные числа: \n', eig_values)
print('Собственные векторы: \n', eig_vectors)
```

```python
import numpy as np
x1 = np.array([1, 2, 1, 1]).T
x2 = np.array([70, 130, 65, 60]).T

C = np.array([[1, 0.9922], [0.9922, 1]])

eig_values, eig_vectors = np.linalg.eig(C)

x1_norm = (x1 - x1.mean()) / np.linalg.norm(x1)
x2_norm = (x2 - x2.mean()) / np.linalg.norm(x2)

x_new = x1_norm * eig_vectors[0][0] + x2_norm * eig_vectors[1][0]

x_new_norm = (x_new - x_new.mean()) / np.linalg.norm(x_new)

print(np.round(x_new_norm, 2))

## [-0.24  0.86 -0.29 -0.33]
```
