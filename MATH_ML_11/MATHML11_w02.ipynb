```python
import numpy as np
import pandas as pd
A = np.matrix([[1,2,3,4],
               [5,5,6,7],
               [1,4,2,3],
               [5,3,2,1],
               [8,1,2,2]])

df = pd.DataFrame(A,columns  = ['x1','x2','x3','x4'])
df_std  = (df - df.mean()) / (df.std())

df_std
```

```python
cov_mat = np.cov(df_std.T)
cov_mat
```

```python
eigen_val, eigen_vectors = np.linalg.eig(cov_mat)

display(eigen_val)
display(eigen_vectors)
```

–¢–µ–ø–µ—Ä—å –Ω–∞–º –Ω–µ–æ–±—Ö–æ–¥–∏–º–æ –≤—ã–±—Ä–∞—Ç—å –¥–≤–∞ —Å–æ–±—Å—Ç–≤–µ–Ω–Ω—ã—Ö –≤–µ–∫—Ç–æ—Ä–∞, —Ç–∞–∫ –∫–∞–∫ –≤ –∏—Ç–æ–≥–µ –º—ã —Ö–æ—Ç–∏–º –ø–æ–ª—É—á–∏—Ç—å –¥–≤–µ –∫–æ–º–ø–æ–Ω–µ–Ω—Ç—ã. –ù—É–∂–Ω–æ –≤–∑—è—Ç—å –≤–µ–∫—Ç–æ—Ä—ã —Å –Ω–∞–∏–±–æ–ª—å—à–∏–º–∏ —Å–æ–±—Å—Ç–≤–µ–Ω–Ω—ã–º–∏ –∑–Ω–∞—á–µ–Ω–∏—è–º–∏. –ü—Ä–∏ –Ω–∞—Ö–æ–∂–¥–µ–Ω–∏–∏ —Å–æ–±—Å—Ç–≤–µ–Ω–Ω—ã—Ö –≤–µ–∫—Ç–æ—Ä–æ–≤ –≤ Python –æ–Ω–∏ –≤—ã–≤–æ–¥—è—Ç—Å—è –≤ —Å–æ–æ—Ç–≤–µ—Ç—Å—Ç–≤–∏–∏ —Å —É–±—ã–≤–∞–Ω–∏–µ–º —Å–æ–æ—Ç–≤–µ—Ç—Å—Ç–≤—É—é—â–∏—Ö —Å–æ–±—Å—Ç–≤–µ–Ω–Ω—ã—Ö –∑–Ω–∞—á–µ–Ω–∏–π, –ø–æ—ç—Ç–æ–º—É –±–µ—Ä—ë–º –ø–µ—Ä–≤—ã–µ –¥–≤–∞ –≤–µ–∫—Ç–æ—Ä–∞:

```python
first_two = eigen_vectors[:, :2]
first_two

#–ö–∞–∫ –≤–∑—è—Ç—å 2 –ø–µ—Ä–≤—ã—Ö –≤–µ–∫—Ç–æ—Ä–∞?
```

```python
Matrix_PC = df_std @ first_two
Matrix_PC
```

–í –ø–∞—Ä–∞–º–µ—Ç—Ä n_components –≤ –∫–∞—á–µ—Å—Ç–≤–µ –∑–Ω–∞—á–µ–Ω–∏—è –º–æ–∂–Ω–æ –ø–µ—Ä–µ–¥–∞—Ç—å –∫–æ–ª–∏—á–µ—Å—Ç–≤–æ –Ω–µ–æ–±—Ö–æ–¥–∏–º—ã—Ö –∫–æ–º–ø–æ–Ω–µ–Ω—Ç–æ–≤ –∏–ª–∏ –º–∏–Ω–∏–º–∞–ª—å–Ω–æ –¥–æ–ø—É—Å—Ç–∏–º—É—é –æ–±—ä—è—Å–Ω—ë–Ω–Ω—É—é –¥–∏—Å–ø–µ—Ä—Å–∏—é –≤ –≤–∏–¥–µ –¥–µ—Å—è—Ç–∏—á–Ω–æ–π –¥—Ä–æ–±–∏. –ù–∞–ø—Ä–∏–º–µ—Ä, –µ—Å–ª–∏ –Ω–∞–º –Ω—É–∂–Ω–æ —Å—Ç–æ–ª—å–∫–æ –∫–æ–º–ø–æ–Ω–µ–Ω—Ç, —á—Ç–æ–±—ã –æ–Ω–∏ –æ–±—ä—è—Å–Ω—è–ª–∏ –Ω–µ –º–µ–Ω–µ–µ 90 % —Ä–∞–∑–±—Ä–æ—Å–∞ –¥–∞–Ω–Ω—ã—Ö, —Ç–æ –º—ã –∑–∞–ø–∏—à–µ–º n_components = 0.9.

```python
#–∏–º–ø–æ—Ä—Ç–∏—Ä—É–µ–º –Ω—É–∂–Ω—ã–π –∞–ª–≥–æ—Ä–∏—Ç–º

from sklearn.decomposition import PCA
#–æ–ø—Ä–µ–¥–µ–ª—è–µ–º –º–µ—Ç–æ–¥ –≥–ª–∞–≤–Ω—ã—Ö –∫–æ–º–ø–æ–Ω–µ–Ω—Ç —Å –¥–≤—É–º—è –∫–æ–º–ø–æ–Ω–µ–Ω—Ç–∞–º–∏
pca = PCA(n_components=2)
#–æ–±—É—á–∞–µ–º –∞–ª–≥–æ—Ä–∏—Ç–º –Ω–∞ –Ω–∞—à–∏—Ö –¥–∞–Ω–Ω—ã—Ö
principalComponents = pca.fit_transform(df_std)

# –≤—ã–≤–æ–¥–∏–º —Ä–µ–∑—É–ª—å—Ç–∞—Ç
principalComponents
display(principalComponents)

# –≤—ã–≤–æ–¥–∏–º –¥–∏—Å–ø–µ—Ä—Å–∏—é
display(pca.explained_variance_)

# –≤—ã–≤–æ–¥–∏–º –¥–æ–ª—é –¥–∏—Å–ø–µ—Ä—Å–∏–∏
display(pca.explained_variance_ratio_)

# –≤—ã–≤–æ–¥–∏–º —Å—É–º–º—É –¥–∏—Å–ø–µ—Ä—Å–∏–π
display(pca.explained_variance_ratio_.cumsum())

```

```python
# –ó–∞–¥–∞–Ω–∏–µ 5.1: –ú–∞—Ç—Ä–∏—Ü–∞ –∫–æ–≤–∞—Ä–∏–∞—Ü–∏–π –¥–ª—è –≤–µ–∫—Ç–æ—Ä–æ–≤ (3, 4, 1) –∏ (1, 6, 2)

# –°–æ–∑–¥–∞—ë–º –≤–µ–∫—Ç–æ—Ä—ã
v1 = np.array([3, 4, 1])
v2 = np.array([1, 6, 2])

# –û–±—ä–µ–¥–∏–Ω—è–µ–º –≤–µ–∫—Ç–æ—Ä—ã –≤ –º–∞—Ç—Ä–∏—Ü—É (–∫–∞–∂–¥–∞—è —Å—Ç—Ä–æ–∫–∞ - —ç—Ç–æ –ø–µ—Ä–µ–º–µ–Ω–Ω–∞—è)
# –î–ª—è np.cov() –Ω—É–∂–Ω–æ –ø–µ—Ä–µ–¥–∞—Ç—å –º–∞—Ç—Ä–∏—Ü—É, –≥–¥–µ –∫–∞–∂–¥–∞—è —Å—Ç—Ä–æ–∫–∞ - –ø–µ—Ä–µ–º–µ–Ω–Ω–∞—è, —Å—Ç–æ–ª–±—Ü—ã - –Ω–∞–±–ª—é–¥–µ–Ω–∏—è
data = np.array([v1, v2])

# –í—ã—á–∏—Å–ª—è–µ–º –º–∞—Ç—Ä–∏—Ü—É –∫–æ–≤–∞—Ä–∏–∞—Ü–∏–π
cov_matrix = np.cov(data)

print("–ú–∞—Ç—Ä–∏—Ü–∞ –∫–æ–≤–∞—Ä–∏–∞—Ü–∏–π:")
print(cov_matrix)
print()

# –ù–∞—Ö–æ–¥–∏–º —Å—É–º–º—É –≤—Å–µ—Ö —ç–ª–µ–º–µ–Ω—Ç–æ–≤ –º–∞—Ç—Ä–∏—Ü—ã
sum_cov = np.sum(cov_matrix)

print(f"–°—É–º–º–∞ –≤—Å–µ—Ö –∑–Ω–∞—á–µ–Ω–∏–π –º–∞—Ç—Ä–∏—Ü—ã: {sum_cov}")
print(f"–°—É–º–º–∞ –≤—Å–µ—Ö –∑–Ω–∞—á–µ–Ω–∏–π –º–∞—Ç—Ä–∏—Ü—ã (–æ–∫—Ä—É–≥–ª–µ–Ω–æ –¥–æ 2 –∑–Ω–∞–∫–æ–≤): {round(sum_cov, 2)}")

# –í—ã–≤–æ–¥–∏–º —Ä–µ–∑—É–ª—å—Ç–∞—Ç
display(cov_matrix)
display(f"–û—Ç–≤–µ—Ç: {round(sum_cov, 2)}")
```

1 point possible (graded)
–î–∞–Ω–∞ –º–∞—Ç—Ä–∏—Ü–∞ –ø—Ä–∏–∑–Ω–∞–∫–æ–≤:
A = np.matrix([[8,7,2,9],
               [1,3,6,3],
               [7,2,0,3],
               [10,3,1,1],
               [8,1,3,4]])

–ö–∞–∫–æ–µ –º–∏–Ω–∏–º–∞–ª—å–Ω–æ–µ –∫–æ–ª–∏—á–µ—Å—Ç–≤–æ –≥–ª–∞–≤–Ω—ã—Ö –∫–æ–º–ø–æ–Ω–µ–Ω—Ç –Ω–∞–¥–æ –≤—ã–¥–µ–ª–∏—Ç—å, —á—Ç–æ–±—ã —Å–æ—Ö—Ä–∞–Ω–∏—Ç—å –∏–Ω—Ñ–æ—Ä–º–∞—Ü–∏—é –æ –∫–∞–∫ –º–∏–Ω–∏–º—É–º 90 % —Ä–∞–∑–±—Ä–æ—Å–∞ –¥–∞–Ω–Ω—ã—Ö?

```python
# –ó–∞–¥–∞–Ω–∏–µ 5.4: –°—Ä–∞–≤–Ω–µ–Ω–∏–µ –ø–æ–¥—Ö–æ–¥–æ–≤ - —Å —Å—Ç–∞–Ω–¥–∞—Ä—Ç–∏–∑–∞—Ü–∏–µ–π –∏ –±–µ–∑

from sklearn.decomposition import PCA

# –°–æ–∑–¥–∞—ë–º –º–∞—Ç—Ä–∏—Ü—É –ø—Ä–∏–∑–Ω–∞–∫–æ–≤
A = np.matrix([[8,7,2,9],
               [1,3,6,3],
               [7,2,0,3],
               [10,3,1,1],
               [8,1,3,4]])

print("=" * 80)
print("–°–†–ê–í–ù–ï–ù–ò–ï –ü–û–î–•–û–î–û–í: –ë–ï–ó –°–¢–ê–ù–î–ê–†–¢–ò–ó–ê–¶–ò–ò vs –°–û –°–¢–ê–ù–î–ê–†–¢–ò–ó–ê–¶–ò–ï–ô")
print("=" * 80)
print()

# ============================================================================
# –ü–û–î–•–û–î 1: –ë–ï–ó –°–¢–ê–ù–î–ê–†–¢–ò–ó–ê–¶–ò–ò (–º–æ–π –ø–µ—Ä–≤–æ–Ω–∞—á–∞–ª—å–Ω—ã–π –ø–æ–¥—Ö–æ–¥)
# ============================================================================
print("–ü–û–î–•–û–î 1: –ë–ï–ó –°–¢–ê–ù–î–ê–†–¢–ò–ó–ê–¶–ò–ò")
print("-" * 80)

df_A = pd.DataFrame(A)

pca_no_std = PCA(n_components=0.9)
pca_no_std.fit(df_A)

min_components_no_std = pca_no_std.n_components_
print(f"–ú–∏–Ω–∏–º–∞–ª—å–Ω–æ–µ –∫–æ–ª–∏—á–µ—Å—Ç–≤–æ –∫–æ–º–ø–æ–Ω–µ–Ω—Ç: {min_components_no_std}")
print(f"–û–±—ä—è—Å–Ω—ë–Ω–Ω–∞—è –¥–∏—Å–ø–µ—Ä—Å–∏—è: {pca_no_std.explained_variance_ratio_.sum():.4f} ({pca_no_std.explained_variance_ratio_.sum()*100:.2f}%)")

# –î–µ—Ç–∞–ª—å–Ω—ã–π –∞–Ω–∞–ª–∏–∑
pca_all_no_std = PCA(n_components=None)
pca_all_no_std.fit(df_A)
cumsum_no_std = np.cumsum(pca_all_no_std.explained_variance_ratio_)

print("\n–î–æ–ª–∏ –æ–±—ä—è—Å–Ω—ë–Ω–Ω–æ–π –¥–∏—Å–ø–µ—Ä—Å–∏–∏ (–ë–ï–ó —Å—Ç–∞–Ω–¥–∞—Ä—Ç–∏–∑–∞—Ü–∏–∏):")
for i, (var_ratio, cum_var) in enumerate(zip(pca_all_no_std.explained_variance_ratio_, cumsum_no_std), 1):
    print(f"  –ö–æ–º–ø–æ–Ω–µ–Ω—Ç–∞ {i}: {var_ratio:.4f} ({var_ratio*100:.2f}%), –∫—É–º—É–ª—è—Ç–∏–≤–Ω–æ: {cum_var:.4f} ({cum_var*100:.2f}%)")

print()
print("=" * 80)

# ============================================================================
# –ü–û–î–•–û–î 2: –°–û –°–¢–ê–ù–î–ê–†–¢–ò–ó–ê–¶–ò–ï–ô (—ç—Ç–∞–ª–æ–Ω–Ω–æ–µ —Ä–µ—à–µ–Ω–∏–µ)
# ============================================================================
print("–ü–û–î–•–û–î 2: –°–û –°–¢–ê–ù–î–ê–†–¢–ò–ó–ê–¶–ò–ï–ô (—ç—Ç–∞–ª–æ–Ω–Ω–æ–µ —Ä–µ—à–µ–Ω–∏–µ)")
print("-" * 80)

df = pd.DataFrame(A)
df_std = (df - df.mean()) / (df.std())

print("–°—Ç–∞–Ω–¥–∞—Ä—Ç–∏–∑–∏—Ä–æ–≤–∞–Ω–Ω—ã–µ –¥–∞–Ω–Ω—ã–µ:")
print(df_std)
print()

pca_with_std = PCA(n_components=0.9)
pca_with_std.fit(df_std)

min_components_with_std = pca_with_std.n_components_
print(f"–ú–∏–Ω–∏–º–∞–ª—å–Ω–æ–µ –∫–æ–ª–∏—á–µ—Å—Ç–≤–æ –∫–æ–º–ø–æ–Ω–µ–Ω—Ç: {min_components_with_std}")
print(f"–û–±—ä—è—Å–Ω—ë–Ω–Ω–∞—è –¥–∏—Å–ø–µ—Ä—Å–∏—è: {pca_with_std.explained_variance_ratio_.sum():.4f} ({pca_with_std.explained_variance_ratio_.sum()*100:.2f}%)")

# –î–µ—Ç–∞–ª—å–Ω—ã–π –∞–Ω–∞–ª–∏–∑
pca_all_with_std = PCA(n_components=None)
pca_all_with_std.fit(df_std)
cumsum_with_std = np.cumsum(pca_all_with_std.explained_variance_ratio_)

print("\n–î–æ–ª–∏ –æ–±—ä—è—Å–Ω—ë–Ω–Ω–æ–π –¥–∏—Å–ø–µ—Ä—Å–∏–∏ (–°–û —Å—Ç–∞–Ω–¥–∞—Ä—Ç–∏–∑–∞—Ü–∏–µ–π):")
for i, (var_ratio, cum_var) in enumerate(zip(pca_all_with_std.explained_variance_ratio_, cumsum_with_std), 1):
    print(f"  –ö–æ–º–ø–æ–Ω–µ–Ω—Ç–∞ {i}: {var_ratio:.4f} ({var_ratio*100:.2f}%), –∫—É–º—É–ª—è—Ç–∏–≤–Ω–æ: {cum_var:.4f} ({cum_var*100:.2f}%)")

print()
print("=" * 80)
print("–ò–¢–û–ì–û–í–û–ï –°–†–ê–í–ù–ï–ù–ò–ï:")
print("=" * 80)
print(f"–ë–ï–ó —Å—Ç–∞–Ω–¥–∞—Ä—Ç–∏–∑–∞—Ü–∏–∏: {min_components_no_std} –∫–æ–º–ø–æ–Ω–µ–Ω—Ç(—ã) –¥–ª—è 90% –¥–∏—Å–ø–µ—Ä—Å–∏–∏")
print(f"–°–û —Å—Ç–∞–Ω–¥–∞—Ä—Ç–∏–∑–∞—Ü–∏–µ–π: {min_components_with_std} –∫–æ–º–ø–æ–Ω–µ–Ω—Ç(—ã) –¥–ª—è 90% –¥–∏—Å–ø–µ—Ä—Å–∏–∏")
print()
print("=" * 80)
print("–í–´–í–û–î:")
print("=" * 80)
print("""
–í PCA —Å—Ç–∞–Ω–¥–∞—Ä—Ç–∏–∑–∞—Ü–∏—è –¥–∞–Ω–Ω—ã—Ö –ö–†–ò–¢–ò–ß–ï–°–ö–ò –í–ê–ñ–ù–ê, –ø–æ—Ç–æ–º—É —á—Ç–æ:

1. PCA –æ—Å–Ω–æ–≤–∞–Ω –Ω–∞ –∫–æ–≤–∞—Ä–∏–∞—Ü–∏–æ–Ω–Ω–æ–π –º–∞—Ç—Ä–∏—Ü–µ, –∫–æ—Ç–æ—Ä–∞—è —á—É–≤—Å—Ç–≤–∏—Ç–µ–ª—å–Ω–∞ –∫ –º–∞—Å—à—Ç–∞–±—É –ø—Ä–∏–∑–Ω–∞–∫–æ–≤
2. –ü—Ä–∏–∑–Ω–∞–∫–∏ —Å –±–æ–ª—å—à–∏–º–∏ –∑–Ω–∞—á–µ–Ω–∏—è–º–∏ –±—É–¥—É—Ç –¥–æ–º–∏–Ω–∏—Ä–æ–≤–∞—Ç—å –≤ –∞–Ω–∞–ª–∏–∑–µ
3. –ë–µ–∑ —Å—Ç–∞–Ω–¥–∞—Ä—Ç–∏–∑–∞—Ü–∏–∏ PCA –±—É–¥–µ—Ç –∏—Å–∫–∞—Ç—å –Ω–∞–ø—Ä–∞–≤–ª–µ–Ω–∏—è –º–∞–∫—Å–∏–º–∞–ª—å–Ω–æ–π –¥–∏—Å–ø–µ—Ä—Å–∏–∏ –≤ –∏—Å—Ö–æ–¥–Ω–æ–º –º–∞—Å—à—Ç–∞–±–µ,
   —á—Ç–æ –º–æ–∂–µ—Ç –Ω–µ –æ—Ç—Ä–∞–∂–∞—Ç—å —Ä–µ–∞–ª—å–Ω—É—é —Å—Ç—Ä—É–∫—Ç—É—Ä—É –¥–∞–Ω–Ω—ã—Ö

–≠–¢–ê–õ–û–ù–ù–û–ï –†–ï–®–ï–ù–ò–ï –ü–†–ê–í–ò–õ–¨–ù–û–ï - –Ω—É–∂–Ω–æ –∏—Å–ø–æ–ª—å–∑–æ–≤–∞—Ç—å —Å—Ç–∞–Ω–¥–∞—Ä—Ç–∏–∑–∞—Ü–∏—é!
""")

# –í—ã–≤–æ–¥–∏–º –∏—Ç–æ–≥–æ–≤—ã–π –æ—Ç–≤–µ—Ç
display(Markdown(f"**‚úÖ –ü–†–ê–í–ò–õ–¨–ù–´–ô –û–¢–í–ï–¢: –º–∏–Ω–∏–º–∞–ª—å–Ω–æ–µ –∫–æ–ª–∏—á–µ—Å—Ç–≤–æ –≥–ª–∞–≤–Ω—ã—Ö –∫–æ–º–ø–æ–Ω–µ–Ω—Ç = {min_components_with_std}**"))
display(Markdown(f"*(–ø—Ä–∏ –∏—Å–ø–æ–ª—å–∑–æ–≤–∞–Ω–∏–∏ —Å—Ç–∞–Ω–¥–∞—Ä—Ç–∏–∑–∞—Ü–∏–∏ –¥–∞–Ω–Ω—ã—Ö)*"))
```
