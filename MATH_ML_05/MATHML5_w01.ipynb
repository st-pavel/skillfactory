$f(x) = -2x^3 -3x^2 + 12x - 4$

В качестве ответа запишите точку минимума:

```python
from sympy import Symbol, diff, solveset, Eq

x = Symbol('x')

f = -2*x**3 -3*x**2 + 12*x - 4
f_prime = (diff(f, x))

x_extremum = solveset(Eq(f_prime, 0), x)

display(Markdown(f"""
### Функция: $f(x) = {latex(f)}$

### Производная: $f'(x) = {latex(f_prime)}$

### Точки экстремума: $x = {latex(x_extremum)}$
"""))


# Находим вторую производную
f_double_prime = diff(f_prime, x)

# Создаем список для хранения точек минимума
minima = []

# Проверяем каждую точку экстремума
for point in x_extremum:
    # Вычисляем значение второй производной в точке
    second_derivative_value = f_double_prime.subs(x, point)
    
    # Если вторая производная положительная - это минимум
    if second_derivative_value > 0:
        minima.append(point)

print(f"Точки минимума функции: {minima}")

# Если нужно найти значение функции в точках минимума:
for point in minima:
    value = f.subs(x, point)
    print(f"Значение функции в точке {point}: {value}")
   
```

### Задание 3.3
1 из 1 балла (оценивается)
Подставьте в полученную матрицу Гессе каждую из точек и определите их характер:

```python
x = Symbol('x')
y = Symbol('y')

z = x**3 + 3*x*y**2 -15*x - 12*y + 1

z_prime_x = diff(z, x)
z_prime_y = diff(z, y)
display(Markdown(f"""
### Функция: $z = {latex(z)}$

### Производная: $z'_x = {latex(z_prime_x)}$

### Производная: $z'_y = {latex(z_prime_y)}$
"""))

# Найдем экстремумы

x_extremum = solveset(Eq(z_prime_x, 0), x)
y_extremum = solveset(Eq(z_prime_y, 0), y)

eq_x = Eq(z_prime_x, 0)
eq_y = Eq(z_prime_y, 0)

sol = solve([eq_x, eq_y], [x, y])

display(Markdown(f"""
### Точки экстремума: $x = {latex(sol)}$

"""))

hessian = Matrix([[diff(diff(z, x), x), diff(diff(z, x), y)],
                  [diff(diff(z, y), x), diff(diff(z, y), y)]])

display(Markdown(f"""
### Матрица Гессе: $H = {latex(hessian)}$

***
"""))

# Подставим точки экстремума в матрицу Гессе

for point in sol:
    hessian_value = hessian.subs({x: point[0], y: point[1]})
    # Найдем определитель матрицы Гессе в точке
    det_hessian = hessian_value.det()
    
    # # Найдем собственные значения матрицы Гессе в точке    
    # Определим тип экстремума через собственные значения
    # если одно собственное значение равно 0, то это неопределённость
    
    # Получаем собственные значения и векторы
    eigen_data = hessian_value.eigenvects()
    
    # Собственные значения
    eigen_values = [entry[0] for entry in eigen_data]
    
    # Собственные векторы
    eigen_vectors = [entry[2][0] for entry in eigen_data]
       
        
    # Проверяем тип экстремума
    if 0 in eigen_values:
        extremum_type = "граничный случай, тест Гессе не даёт однозначного ответа"
    elif all(float(ev) > 0 for ev in eigen_values):
        extremum_type = "минимум"
    elif all(float(ev) < 0 for ev in eigen_values):
        extremum_type = "максимум"
    else:
        extremum_type = "седловая точка"
        
    display(Markdown(f"""
### Матрица Гессе в точке {point} : $H = {latex(hessian_value)}$

### Определитель матрицы Гессе в точке {point} : $det(H) = {latex(det_hessian)}$

### Собственные значения матрицы Гессе в точке {point} : $eigenvalues(H) = {latex(eigen_values)}$

### Собственные векторы матрицы Гессе в точке {point} : $eigenvectors(H) = {latex(eigen_vectors)}$

### Тип экстремума в точке {point} : {extremum_type}

***
"""))
    
```
