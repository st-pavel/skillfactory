–ù–µ–æ–±—Ö–æ–¥–∏–º–æ –∫–ª–∞—Å—Å–∏—Ñ–∏—Ü–∏—Ä–æ–≤–∞—Ç—å —Å–µ–º–µ–π—Å—Ç–≤–∞ –ª—è–≥—É—à–µ–∫ –≤ –∑–∞–≤–∏—Å–∏–º–æ—Å—Ç–∏ –æ—Ç –æ—Å–æ–±–µ–Ω–Ω–æ—Å—Ç–µ–π –∏—Ö –∫–≤–∞–∫–∞–Ω—å—è –∏ –ø—Ä–æ—á–∏—Ö –∑–≤—É–∫–æ–≤—ã—Ö —ç—Ñ—Ñ–µ–∫—Ç–æ–≤.

–ë—É–¥–µ–º —Ä–µ—à–∞—Ç—å –∑–∞–¥–∞—á—É –±–∏–Ω–∞—Ä–Ω–æ–π –∫–ª–∞—Å—Å–∏—Ñ–∏–∫–∞—Ü–∏–∏ –ø–æ –≤—ã—è–≤–ª–µ–Ω–∏—é –ª—è–≥—É—à–µ–∫, –∫–æ—Ç–æ—Ä—ã–µ –æ—Ç–Ω–æ—Å—è—Ç—Å—è –∫ —Å–µ–º–µ–π—Å—Ç–≤—É 'Dendrobatidae' (–ø—Ä–∏–∑–Ω–∞–∫ 'Family'). –°–µ–º–µ–π—Å—Ç–≤–æ 'Dendrobatidae' –±—É–¥–µ—Ç –∫–ª–∞—Å—Å–æ–º 1, –≤—Å–µ –æ—Å—Ç–∞–ª—å–Ω—ã–µ —Å–µ–º–µ–π—Å—Ç–≤–∞ ‚Äî –∫–ª–∞—Å—Å–æ–º 0.

–í –∫–∞—á–µ—Å—Ç–≤–µ –ø—Ä–∏–∑–Ω–∞–∫–æ–≤, –∫–æ—Ç–æ—Ä—ã–µ –º—ã –±—É–¥–µ–º –∏—Å–ø–æ–ª—å–∑–æ–≤–∞—Ç—å –¥–ª—è –ø—Ä–µ–¥—Å–∫–∞–∑–∞–Ω–∏—è, –Ω–µ–æ–±—Ö–æ–¥–∏–º–æ –≤–∑—è—Ç—å –≤—Å–µ, –∫—Ä–æ–º–µ:

'Family' ‚Äî —Å–µ–º–µ–π—Å—Ç–≤–æ –ª—è–≥—É—à–µ–∫;
'Genus' ‚Äî —Ä–æ–¥ –ª—è–≥—É—à–µ–∫;
'Species' ‚Äî –≤–∏–¥ –ª—è–≥—É—à–µ–∫;
'RecordID' ‚Äî ID –∑–∞–ø–∏—Å–∏.
–í—Å–µ –æ—Å—Ç–∞–ª—å–Ω—ã–µ –ø—Ä–∏–∑–Ω–∞–∫–∏ –æ—Ç–Ω–æ—Å—è—Ç—Å—è –∫ –∞–∫—É—Å—Ç–∏—á–µ—Å–∫–∏–º –æ—Å–æ–±–µ–Ω–Ω–æ—Å—Ç—è–º –∫–≤–∞–∫–∞–Ω—å—è ‚Äî –æ–Ω–∏-—Ç–æ –Ω–∞–º –∏ –ø–æ–Ω–∞–¥–æ–±—è—Ç—Å—è, —á—Ç–æ–±—ã –æ–ø—Ä–µ–¥–µ–ª–∏—Ç—å, –∫ –∫–∞–∫–æ–º—É —Å–µ–º–µ–π—Å—Ç–≤—É –æ—Ç–Ω–æ—Å–∏—Ç—Å—è –ª—è–≥—É—à–∫–∞.

–†–∞–∑–¥–µ–ª–∏—Ç–µ –≤—ã–±–æ—Ä–∫—É –Ω–∞ –æ–±—É—á–∞—é—â—É—é –∏ —Ç–µ—Å—Ç–æ–≤—É—é –≤ —Å–æ–æ—Ç–Ω–æ—à–µ–Ω–∏–∏ 80/20, –ø–∞—Ä–∞–º–µ—Ç—Ä random_state = 31.

```python
# –ü–æ–¥–≥–æ—Ç–æ–≤–∫–∞ –¥–∞–Ω–Ω—ã—Ö –¥–ª—è –±–∏–Ω–∞—Ä–Ω–æ–π –∫–ª–∞—Å—Å–∏—Ñ–∏–∫–∞—Ü–∏–∏

# –°–æ–∑–¥–∞–µ–º —Ü–µ–ª–µ–≤—É—é –ø–µ—Ä–µ–º–µ–Ω–Ω—É—é: Dendrobatidae = 1, –æ—Å—Ç–∞–ª—å–Ω—ã–µ = 0
y = (df['Family'] == 'Dendrobatidae').astype(int)

print(f"–†–∞—Å–ø—Ä–µ–¥–µ–ª–µ–Ω–∏–µ –∫–ª–∞—Å—Å–æ–≤:")
print(y.value_counts())
print(f"\n–î–æ–ª—è –∫–ª–∞—Å—Å–∞ 1 (Dendrobatidae): {y.mean():.4f}")

# –í—ã–±–∏—Ä–∞–µ–º –ø—Ä–∏–∑–Ω–∞–∫–∏: –≤—Å–µ –∫—Ä–æ–º–µ 'Family', 'Genus', 'Species', 'RecordID'
columns_to_exclude = ['Family', 'Genus', 'Species', 'RecordID']
X = df.drop(columns=columns_to_exclude)

print(f"\n–ö–æ–ª–∏—á–µ—Å—Ç–≤–æ –ø—Ä–∏–∑–Ω–∞–∫–æ–≤: {X.shape[1]}")
print(f"–ù–∞–∑–≤–∞–Ω–∏—è –ø—Ä–∏–∑–Ω–∞–∫–æ–≤:")
print(X.columns.tolist())

# –†–∞–∑–¥–µ–ª–µ–Ω–∏–µ –Ω–∞ –æ–±—É—á–∞—é—â—É—é –∏ —Ç–µ—Å—Ç–æ–≤—É—é –≤—ã–±–æ—Ä–∫–∏ 80/20
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(
    X, y, 
    test_size=0.2, 
    random_state=31
    #stratify=y  # –°–æ—Ö—Ä–∞–Ω—è–µ–º –ø—Ä–æ–ø–æ—Ä—Ü–∏–∏ –∫–ª–∞—Å—Å–æ–≤
)

print(f"\n–†–∞–∑–º–µ—Ä –æ–±—É—á–∞—é—â–µ–π –≤—ã–±–æ—Ä–∫–∏: {X_train.shape}")
print(f"–†–∞–∑–º–µ—Ä —Ç–µ—Å—Ç–æ–≤–æ–π –≤—ã–±–æ—Ä–∫–∏: {X_test.shape}")
print(f"\n–†–∞—Å–ø—Ä–µ–¥–µ–ª–µ–Ω–∏–µ –∫–ª–∞—Å—Å–æ–≤ –≤ –æ–±—É—á–∞—é—â–µ–π –≤—ã–±–æ—Ä–∫–µ:")
print(y_train.value_counts())
print(f"\n–†–∞—Å–ø—Ä–µ–¥–µ–ª–µ–Ω–∏–µ –∫–ª–∞—Å—Å–æ–≤ –≤ —Ç–µ—Å—Ç–æ–≤–æ–π –≤—ã–±–æ—Ä–∫–µ:")
print(y_test.value_counts())
```

## –ó–∞–¥–∞–Ω–∏–µ 7.5
1 point possible (graded)
–î–ª—è –Ω–∞—á–∞–ª–∞ –æ–±—É—á–∏—Ç–µ –Ω–∞ –Ω–∞—à–∏—Ö –¥–∞–Ω–Ω—ã—Ö —Å–ª—É—á–∞–π–Ω—ã–π –ª–µ—Å —Å –¥–µ—Å—è—Ç—å—é —Ä–µ—à–∞—é—â–∏–º–∏ –¥–µ—Ä–µ–≤—å—è–º–∏. –í–æ—Å–ø–æ–ª—å–∑—É–π—Ç–µ—Å—å –ø–∞—Ä–∞–º–µ—Ç—Ä–∞–º–∏ –ø–æ —É–º–æ–ª—á–∞–Ω–∏—é. –í –∫–∞—á–µ—Å—Ç–≤–µ –∑–Ω–∞—á–µ–Ω–∏—è random_state –≤–æ–∑—å–º–∏—Ç–µ —á–∏—Å–ª–æ 42.

–û—Ü–µ–Ω–∏—Ç–µ –∑–Ω–∞—á–µ–Ω–∏–µ F1-–º–µ—Ä—ã –∏ –≤–≤–µ–¥–∏—Ç–µ –µ–≥–æ –≤ –∫–∞—á–µ—Å—Ç–≤–µ –æ—Ç–≤–µ—Ç–∞, –ø—Ä–µ–¥–≤–∞—Ä–∏—Ç–µ–ª—å–Ω–æ –æ–∫—Ä—É–≥–ª–∏–≤ –¥–æ –¥–≤—É—Ö –∑–Ω–∞–∫–æ–≤ –ø–æ—Å–ª–µ —Ç–æ—á–∫–∏-—Ä–∞–∑–¥–µ–ª–∏—Ç–µ–ª—è.

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import f1_score
model_rfc = RandomForestClassifier(10, random_state=42)
model_rfc.fit(X_train, y_train)
pred = model_rfc.predict(X_test)
f1_score(y_test, pred)
```

```python
# –ó–∞–¥–∞–Ω–∏–µ 7.5: –û–±—É—á–µ–Ω–∏–µ —Å–ª—É—á–∞–π–Ω–æ–≥–æ –ª–µ—Å–∞ —Å 10 –¥–µ—Ä–µ–≤—å—è–º–∏

from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import f1_score

# –°–æ–∑–¥–∞–µ–º –º–æ–¥–µ–ª—å —Å–ª—É—á–∞–π–Ω–æ–≥–æ –ª–µ—Å–∞ —Å 10 –¥–µ—Ä–µ–≤—å—è–º–∏
# –ü–∞—Ä–∞–º–µ—Ç—Ä—ã –ø–æ —É–º–æ–ª—á–∞–Ω–∏—é, random_state = 42
rf_model = RandomForestClassifier(n_estimators=10, random_state=42)

# –û–±—É—á–∞–µ–º –º–æ–¥–µ–ª—å –Ω–∞ –æ–±—É—á–∞—é—â–µ–π –≤—ã–±–æ—Ä–∫–µ
rf_model.fit(X_train, y_train)

# –î–µ–ª–∞–µ–º –ø—Ä–µ–¥—Å–∫–∞–∑–∞–Ω–∏—è –Ω–∞ —Ç–µ—Å—Ç–æ–≤–æ–π –≤—ã–±–æ—Ä–∫–µ
y_pred = rf_model.predict(X_test)

# –í—ã—á–∏—Å–ª—è–µ–º F1-–º–µ—Ä—É
f1 = f1_score(y_test, y_pred)

# –û–∫—Ä—É–≥–ª—è–µ–º –¥–æ 2 –∑–Ω–∞–∫–æ–≤ –ø–æ—Å–ª–µ –∑–∞–ø—è—Ç–æ–π
f1_rounded = round(f1, 2)

print(f"F1-–º–µ—Ä–∞: {f1:.4f}")
print(f"F1-–º–µ—Ä–∞ (–æ–∫—Ä—É–≥–ª–µ–Ω–æ –¥–æ 2 –∑–Ω–∞–∫–æ–≤): {f1_rounded}")
```

–ü—Ä–∞–≤–∏–ª—å–Ω—ã–π —ç—Ç–∞–ª–æ–Ω–Ω—ã–π –æ—Ç–≤–µ—Ç 0.97

## –ó–∞–¥–∞–Ω–∏–µ 7.6
1 point possible (graded)
–¢–µ–ø–µ—Ä—å –ø–æ–ø—Ä–æ–±—É–µ–º —É–ª—É—á—à–∏—Ç—å –∫–∞—á–µ—Å—Ç–≤–æ –Ω–∞—à–µ–≥–æ –ø—Ä–µ–¥—Å–∫–∞–∑–∞–Ω–∏—è –∑–∞ —Å—á—ë—Ç –∏—Å–ø–æ–ª—å–∑–æ–≤–∞–Ω–∏—è —Å—Ç–µ–∫–∏–Ω–≥–∞.

–í –∫–∞—á–µ—Å—Ç–≤–µ –±–∞–∑–æ–≤—ã—Ö –º–æ–¥–µ–ª–µ–π –≤—ã–±–µ—Ä–∏—Ç–µ —Å–ª–µ–¥—É—é—â–∏–µ:

—Å–ª—É—á–∞–π–Ω—ã–π –ª–µ—Å —Å –¥–µ—Å—è—Ç—å—é –¥–µ—Ä–µ–≤—å—è–º–∏, random_state = 31;
KNN, –∫–æ–ª–∏—á–µ—Å—Ç–≤–æ —Å–æ—Å–µ–¥–µ–π = 11;
–Ω–∞–∏–≤–Ω—ã–π –±–∞–π–µ—Å–æ–≤—Å–∫–∏–π –∫–ª–∞—Å—Å–∏—Ñ–∏–∫–∞—Ç–æ—Ä —Å –ø–∞—Ä–∞–º–µ—Ç—Ä–∞–º–∏ –ø–æ —É–º–æ–ª—á–∞–Ω–∏—é, –≤ –∫–∞—á–µ—Å—Ç–≤–µ –º–µ—Ç–æ–¥–∞ –≤–æ–∑—å–º–∏—Ç–µ GaussianNB().
–í –∫–∞—á–µ—Å—Ç–≤–µ –º–µ—Ç–∞–º–æ–¥–µ–ª–∏ –≤—ã–±–µ—Ä–∏—Ç–µ –ª–æ–≥–∏—Å—Ç–∏—á–µ—Å–∫—É—é —Ä–µ–≥—Ä–µ—Å—Å–∏—é.

–û–±—É—á–∏—Ç–µ –º–æ–¥–µ–ª–∏ –∏ —Å–¥–µ–ª–∞–π—Ç–µ –ø—Ä–µ–¥—Å–∫–∞–∑–∞–Ω–∏–µ —Ü–µ–ª–µ–≤–æ–π –º–µ—Ç–∫–∏ –¥–ª—è —Ç–µ—Å—Ç–æ–≤–æ–≥–æ –Ω–∞–±–æ—Ä–∞ –¥–∞–Ω–Ω—ã—Ö.

–†–∞—Å—Å—á–∏—Ç–∞–π—Ç–µ F1-–º–µ—Ä—É –¥–ª—è —Ç–µ—Å—Ç–æ–≤–æ–≥–æ –Ω–∞–±–æ—Ä–∞ –¥–∞–Ω–Ω—ã—Ö –∏ –≤–≤–µ–¥–∏—Ç–µ –µ—ë –≤ –∫–∞—á–µ—Å—Ç–≤–µ –æ—Ç–≤–µ—Ç–∞, –ø—Ä–µ–¥–≤–∞—Ä–∏—Ç–µ–ª—å–Ω–æ –æ–∫—Ä—É–≥–ª–∏–≤ –¥–æ –¥–≤—É—Ö –∑–Ω–∞–∫–æ–≤ –ø–æ—Å–ª–µ —Ç–æ—á–∫–∏-—Ä–∞–∑–¥–µ–ª–∏—Ç–µ–ª—è.

```python
# –ó–∞–¥–∞–Ω–∏–µ 7.6: –°—Ç–µ–∫–∏–Ω–≥ —Å –Ω–µ—Å–∫–æ–ª—å–∫–∏–º–∏ –±–∞–∑–æ–≤—ã–º–∏ –º–æ–¥–µ–ª—è–º–∏

from sklearn.ensemble import RandomForestClassifier, StackingClassifier
from sklearn.neighbors import KNeighborsClassifier
from sklearn.naive_bayes import GaussianNB
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import f1_score

# –°–æ–∑–¥–∞–µ–º –±–∞–∑–æ–≤—ã–µ –º–æ–¥–µ–ª–∏
base_models = [
    ('rf', RandomForestClassifier(n_estimators=10, random_state=31)),
    ('knn', KNeighborsClassifier(n_neighbors=11)),
    ('nb', GaussianNB())
]

# –°–æ–∑–¥–∞–µ–º –º–µ—Ç–∞–º–æ–¥–µ–ª—å (–ª–æ–≥–∏—Å—Ç–∏—á–µ—Å–∫–∞—è —Ä–µ–≥—Ä–µ—Å—Å–∏—è)
meta_model = LogisticRegression()

# –°–æ–∑–¥–∞–µ–º –º–æ–¥–µ–ª—å —Å—Ç–µ–∫–∏–Ω–≥–∞
stacking_model = StackingClassifier(
    estimators=base_models,
    final_estimator=meta_model,
    cv=5  # –ò—Å–ø–æ–ª—å–∑—É–µ–º –∫—Ä–æ—Å—Å-–≤–∞–ª–∏–¥–∞—Ü–∏—é –¥–ª—è –æ–±—É—á–µ–Ω–∏—è –º–µ—Ç–∞–º–æ–¥–µ–ª–∏
)

# –û–±—É—á–∞–µ–º –º–æ–¥–µ–ª—å —Å—Ç–µ–∫–∏–Ω–≥–∞
print("–û–±—É—á–∞–µ–º –º–æ–¥–µ–ª—å —Å—Ç–µ–∫–∏–Ω–≥–∞...")
stacking_model.fit(X_train, y_train)

# –î–µ–ª–∞–µ–º –ø—Ä–µ–¥—Å–∫–∞–∑–∞–Ω–∏—è –Ω–∞ —Ç–µ—Å—Ç–æ–≤–æ–π –≤—ã–±–æ—Ä–∫–µ
y_pred_stacking = stacking_model.predict(X_test)

# –í—ã—á–∏—Å–ª—è–µ–º F1-–º–µ—Ä—É
f1_stacking = f1_score(y_test, y_pred_stacking)

# –û–∫—Ä—É–≥–ª—è–µ–º –¥–æ 2 –∑–Ω–∞–∫–æ–≤ –ø–æ—Å–ª–µ –∑–∞–ø—è—Ç–æ–π
f1_stacking_rounded = round(f1_stacking, 2)

print(f"\nF1-–º–µ—Ä–∞ (—Å—Ç–µ–∫–∏–Ω–≥): {f1_stacking:.4f}")
print(f"F1-–º–µ—Ä–∞ (—Å—Ç–µ–∫–∏–Ω–≥, –æ–∫—Ä—É–≥–ª–µ–Ω–æ –¥–æ 2 –∑–Ω–∞–∫–æ–≤): {f1_stacking_rounded}")
```

–ü—Ä–∞–≤–∏–ª—å–Ω—ã–π —ç—Ç–∞–ª–æ–Ω–Ω—ã–π –æ—Ç–≤–µ—Ç 0.99

```python
estimators = [('rf', RandomForestClassifier(n_estimators=10, random_state=31)),
              ('knn', KNeighborsClassifier(n_neighbors= 11)),
              ('nb', GaussianNB() )               
]
metamodel = StackingClassifier(
    estimators=estimators, final_estimator=LogisticRegression()
)
metamodel.fit(X_train, y_train)
pred = metamodel.predict(X_test)
f1_score(y_test, pred)
```
