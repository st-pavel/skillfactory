### Задание 1.9

1 point possible (graded)

Сколько нетривиальных решений имеет представленная ниже система уравнений?

5w1+w2=0
10w1+2w2=0

Бесконечное множество решений

Перед нами однородная СЛАУ. Такие СЛАУ имеют только два случая решений:

Бесконечное множество решений, если уравнения являются пропорциональными другу, то есть ранг матрицы системы  A  меньше количества неизвестных.
Не имеет решений, если уравнения не являются пропорциональными, то есть ранг матрицы системы  A  равен количеству неизвестных.
Видно, что уравнения в нашей системе пропорциональны друг другу: если умножить первое на 2, получится второе уравнение, а значит одно из них можно убрать. Тогда останется только одно уравнение на две неизвестных:

5w1+w2=0 

Таким образом, количество неизвестных (2) больше, чем количество линейно независимых уравнений, или ранг матрицы (1), а значит система имеет бесконечное множество решений.

```python
v1 = np.array([5, 10])

v2 = np.array([1, 2])

A = np.array([v1,v2]).T

print('A        :\n',A)

print('det(A):\n',np.linalg.det(A))
print('rk(A):\n',np.linalg.matrix_rank(A))
```

### Задание 2.9

1 point possible (graded)

Вычислите вектор ошибок  e  для приближённого решения системы уравнений:
![alt text](2025-04-04_00-24-55.png)


```python
import numpy as np

# Определяем матрицу A и вектор b
A = np.array([[1, -5],
              [2, 1],
              [1, 1]])
b = np.array([1, 2, 2])

# Приближенное решение
w = np.array([1, 1])

# Вычисляем b'
b_prime = A @ w

# Вычисляем вектор ошибок e
e = b - b_prime

print(f"Матрица A:\n{A}")
print(f"Вектор b:\n{b}")
print(f"Приближенное решение w:\n{w}")
print(f"Вычисленное b' = A * w:\n{b_prime}")
print(f"Вектор ошибок e = b - b':\n{e}")
```

### Задание 2.11

7/7 points (graded)

Найдите OLS-оценку для коэффициентов w1, w2 СЛАУ:

![alt text](2025-04-04_08-42-32.png)

```python
import numpy as np
from fractions import Fraction

# Определяем матрицу A и вектор b
A = np.array([[1, 2],
              [-3, 1],
              [1, 2],
              [1, -1]])
b = np.array([1, 4, 5, 0])

# 1. Вычисляем A^T A
ATA = A.T @ A
print("1. Матрица A^T A:")
print(ATA)

# 2. Вычисляем (A^T A)^(-1)
ATA_inv = np.linalg.inv(ATA)
print("\n2. Матрица (A^T A)^(-1):")
print(ATA_inv)

# 3. Вычисляем A^T b
ATb = A.T @ b
print("\nA^T b:")
print(ATb)

# 4. Вычисляем OLS-оценку w = (A^T A)^(-1) * (A^T b)
w = ATA_inv @ ATb
print("\nOLS-оценка w:")
print(w)

# 5. Преобразуем w в дроби и выводим
w_fractions = [Fraction(x).limit_denominator() for x in w]
print(f"\nOLS-оценка в дробях: w1 = {w_fractions[0]}, w2 = {w_fractions[1]}")
```
