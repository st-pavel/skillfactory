### Задача 2.6

1/1 point (graded)

Какая из формул выражает пересечение двух множеств  A∩B  через операции дополнения и объединения?

- $\overline{A \cup B}$
- $A \backslash(B \backslash A)$
- $\overline{\bar{A} \cup \bar{B}}$
- $\overline{(A \cup B)} \cup A$

Ответ: **$\overline{\bar{A} \cup \bar{B}}$**

### Решение:

Это выражение выводится с использованием **законов де Моргана** и свойства двойного отрицания.

1.  Согласно закону де Моргана, дополнение от объединения дополнений равно пересечению множеств:
    $\overline{A} \cup \overline{B} = \overline{A \cap B}$
2.  Применим это к нашему выражению, тогда $\bar{A} \cup \bar{B} = \overline{A \cap B}$
3.  Теперь возьмем дополнение от всего выражения:
    $\overline{\bar{A} \cup \bar{B}} = \overline{\overline{A \cap B}}$
4.  Двойное дополнение ($\overline{\overline{Z}}$) равно самому множеству $Z$. Следовательно:
    $\overline{\overline{A \cap B}} = A \cap B$


Разберем остальные варианты, чтобы убедиться, что они не подходят.

---

#### 1. $\overline{A \cup B}$
*   **Что это значит:** Это «**НЕ** (А или Б)».
*   **Расшифровка:** Сначала мы берем объединение $A \cup B$ (оба круга вместе), а черта сверху берет всё, что находится **снаружи** этих кругов.
*   **Результат:** Это всё внешнее пространство (фон), где нет ни $A$, ни $B$.
*   **Итог:** $\neq A \cap B$.

#### 2. $A \setminus (B \setminus A)$
*   **Что это значит:** Из множества $A$ вычесть «чистую часть $B$».
*   **Пошагово:**
    1.  $B \setminus A$ — это элементы, которые есть в $B$, но которых нет в $A$ .
    2.  Теперь мы берем целое множество $A$ и пытаемся вычесть из него результат шага 1.
    3.  Но множество $A$ и множество «$B$ без $A$» вообще не пересекаются! У них нет общих элементов.
    4.  Если из $A$ вычесть то, чего в нем и так нет, останется просто само $A$.
*   **Результат:** **$A \setminus (B \setminus A)$ = $A$**.
*   **Итог:** $\neq A \cap B$.

#### 4. $\overline{(A \cup B)} \cup A$
*   **Что это значит:** Объединить «всё, что снаружи кругов» и «круг $A$».
*   **Пошагово:**
    1.  $\overline{A \cup B}$ — это внешний фон (см. пункт 1).
    2.  Добавляем к этому фону весь круг $A$.
    3.  Получается всё пространство, **кроме** той части $B$, которая не пересекается с $A$ (лунка от $B$).
*   **Результат:** $A \cup \overline{B}$ (всё $A$ плюс всё, что не $B$).
*   **Итог:** $\neq A \cap B$.

---

### Вывод:
Единственная формула, которая с помощью законов де Моргана сворачивается в пересечение — это  вариант:
$$ \overline{\bar{A} \cup \bar{B}} = \overline{\overline{A \cap B}} = A \cap B $$

### Задача 2.7

1.0/1.0 point (graded)

Пусть A={1,2,3,4}, а B={3,6,5}. 
Выпишите через запятую в порядке возрастания все элементы A∪B, A∩B и A∖B.

### Решение:

Вот результаты операций над множествами $A=\{1, 2, 3, 4\}$ и $B=\{3, 6, 5\}$:

1.  **Объединение ($A \cup B$):**

    Это все элементы, которые есть хотя бы в одном из множеств (без повторений).

    **Ответ:** 1, 2, 3, 4, 5, 6

2.  **Пересечение ($A \cap B$):**

    Это элементы, которые есть **и** в $A$, **и** в $B$ одновременно.

    **Ответ:** 3

3.  **Разность ($A \setminus B$):**

    Это элементы, которые есть в $A$, но которых **нет** в $B$. (Из множества $A$ удаляем тройку, так как она есть в $B$).

    **Ответ:** 1, 2, 4

### Задача 3.1

1/1 point (graded)

Найдите пересечение следующих множеств —  $\{a \in \mathbb{N} \mid n \in \mathbb{N}: a=2 n\}$  и  $\{a \in \mathbb{N} \mid n \in \mathbb{N}: a=3 n\}$ :

**A) $\{a \in \mathbb{N} \mid \exists n \in \mathbb{N} : a = 3n \text{ and } a = 2n\}$**

**B) $\{a \in \mathbb{N} \mid \exists n \in \mathbb{N}, k \in \mathbb{N} : a = 3n \text{ and } a = 2k\}$**

**C) $\{a \in \mathbb{N} \mid \exists n \in \mathbb{N} : a = 6n\}$**

**D) $\{a \in \mathbb{N} \mid \exists n \in \mathbb{N}: \exists k \in \mathbb{N}: a=3 n \text{ or } a=2 k\}$**


Ответ: **B** и **C**.

### Решение:

**1. Пересечение двух множеств — это множество, состоящее из элементов, которые принадлежат ***и первому***, ***и второму*** множеству одновременно.**
*   Первое множество: числа вида $2n$ (четные числа, кратные 2). Примеры: 2, 4, **6**, 8, 10, **12**...
*   Второе множество: числа вида $3n$ (числа, кратные 3). Примеры: 3, **6**, 9, **12**, 15...
*   Пересечение: числа, которые делятся и на 2, и на 3. Наименьшее общее кратное чисел 2 и 3 равно 6. Значит, в пересечении будут числа, кратные 6: **6, 12, 18...**

**2. Анализ вариантов:**

*   ❌ **A) $\{a \in \mathbb{N} \mid \exists n \in \mathbb{N} : a = 3n \text{ and } a = 2n\}$** — **Неверно**.
    Здесь используется одна и та же переменная $n$ для обоих условий. Это означало бы, что число должно быть равно $3n$ и $2n$ при *одном и том же* $n$. Это возможно только если $n=0$ (что дает $a=0$), но для натуральных чисел (1, 2, 3...) таких чисел нет. Например, число 6 получается как $3 \cdot 2$ и $2 \cdot 3$ (множители разные), а эта формула требует одинакового множителя.

*   ✅ **B) $\{a \in \mathbb{N} \mid \exists n \in \mathbb{N}, k \in \mathbb{N} : a = 3n \text{ and } a = 2k\}$** — **Верно**.
    Это правильная **логическая запись** пересечения. Она говорит: число $a$ должно быть кратно 3 (существует какое-то $n$) **И** число $a$ должно быть кратно 2 (существует какое-то $k$). Использование разных букв ($n$ и $k$) позволяет правильно описать условие (например, для числа 6: $n=2$, $k=3$).

*   ✅ **C) $\{a \in \mathbb{N} \mid \exists n \in \mathbb{N} : a = 6n\}$** — **Верно**.
    Это **результат вычисления** пересечения. Как мы выяснили выше, числа, делящиеся на 2 и на 3 одновременно — это числа, делящиеся на 6. Эта формула описывает множество всех чисел, кратных 6. Множества в пунктах B и C полностью совпадают.

*   ❌ **D) $\{... \text{ or } ...\}$** — **Неверно**.
    Слово **or** (или) означает *объединение* множеств (числа, которые делятся или на 2, или на 3), а не их пересечение.

**Итог:**  **B** и **C**, так как оба варианта верно описывают искомое множество.

### Задача 3.2

1/1 point (graded)

Пусть  $A=\{3,6,7\}$ , перечислите через запятую в порядке возрастания элементы множества  $\{n \in \mathbb{Z} \mid n+2 \in A\}$ .



Ответ : **1, 4, 5**.


### Решение:

 Нам нужно найти множество $\{n \in \mathbb{Z} \mid n + 2 \in A\}$.
Это читается так: «все целые числа $n$, для которых выполняется условие: если к $n$ прибавить $2$, результат окажется внутри множества $A$».
Множество $A = \{3, 6, 7\}$.


Таким образом, нам нужно решить уравнение $n + 2 = x$ для каждого элемента $x$ из множества $A$. То есть, чтобы найти $n$, нужно из каждого числа множества $A$ вычесть 2 ($n = x - 2$).

*   Берем первый элемент $3$:
    $n + 2 = 3 \Rightarrow n = 3 - 2 = \mathbf{1}$

*   Берем второй элемент $6$:
    $n + 2 = 6 \Rightarrow n = 6 - 2 = \mathbf{4}$

*   Берем третий элемент $7$:
    $n + 2 = 7 \Rightarrow n = 7 - 2 = \mathbf{5}$

**Итог:**

Получились числа $1, 4, 5$.

В порядке возрастания: **1, 4, 5**.

### Задача 3.3

1/1 point (graded)

Что из нижеперечисленного совпадает с множеством $(A \cap B) \cup(A \backslash B)$?

- $A$
- $B$
- $A \cup B$
- $A \triangle B$



Ответ:  **$A$**.

### Решение:

Для доказательства используем свойства операций над множествами (алгебру множеств).

Выражение:
$$ (A \cap B) \cup (A \setminus B) $$

1.  Разность множеств $A \setminus B$ можно записать как пересечение $A$ с дополнением $B$ (то есть «$A$ и НЕ $B$»):
    $$ A \setminus B = A \cap \overline{B} $$

2.  Подставим это в формулу:
    $$ (A \cap B) \cup (A \cap \overline{B}) $$

3.  При этом,  $A$ есть в обоих частях выражения. Применим **распределительный (дистрибутивный) закон** :
    $$ A \cap (B \cup \overline{B}) $$

4.  Упростим выражение в скобках: $(B \cup \overline{B})$. Объединение множества и всего, что не является этим множеством, дает **Универсум** ($U$) — то есть «всё пространство».
    $$ A \cap U $$

5.  Пересечение любого множества $A$ с универсумом равно самому этому множеству:
    $$ A $$

**Итог:** Выражение упрощается до $A$.
